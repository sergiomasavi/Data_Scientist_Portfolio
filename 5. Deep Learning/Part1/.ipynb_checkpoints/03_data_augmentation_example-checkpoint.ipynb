```python
img_2 = load_img('/content/sample_image.jpg', color_mode='grayscale', target_size=(250,250))
img_2
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img
```

```python
img2_array = img_to_array(img_2)
img2_array.shape
```

```python
np.savez_compressed('img2_comp.npz', img_a=img2_array)
```

```python
img2_dict = np.load('img2_comp.npz')
```

```python
img2_dict['img_a']
```

```python
img2_recovered = array_to_img(img2_dict['img_a'])
img2_recovered
```
