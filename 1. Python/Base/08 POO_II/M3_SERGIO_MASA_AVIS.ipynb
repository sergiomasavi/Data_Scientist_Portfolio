### Creación
* Crear una clase llamada Alumno que tenga los atributos nombre y nota.
* Haremos el constructor de la clase. Añadiremos en el constructor un print para informar de que el alumno se ha creado con éxito.
* Sobreescribiremos el método string (\_\_str_\_\)
* Crearemos un método llamado __calificación__ que imprima por pantalla si el alumno ha aprobado o suspendido en base a su nota. Se extiende a suspendido, aprobado (bien, notable, sobresaliente, matricula de honor).

### Experimentación
* Crearemos algunos alumnos
* Probaremos a mostrar los objetos alumno y el método de calificación

```python
import sys

class Alumno():
    """ 
    Clase alumno.
        Atributos: - nombre. Nombre del alumno (str)
                   - nota. Nota del alumno (float)
    """
    # Atributos de la clase
    nombre = '' # Nombre del alumno 
    nota = 0 # Nota del alumno
    
    # Constructor
    def __init__(self, nombre, nota):
        """ 
        Constructor de la clase 
        
            @param nombre: Nombre del alumno.
            @param nota: Nota del alumno.
        """
        self.nombre = nombre
        
        if nota < 0:
            print('La nota no puede ser inferior a 0. Se tomará este valor por defecto.',sys.stderr)
            self.nota = 0
        elif nota > 10:
            print('La nota no puede ser superior a 10. Se tomará este valor por defecto.',sys.stderr)
            self.nota = 10
        else:
            self.nota = nota
        
        print('Se ha creado correctamente el alumno', self, sep='\n')
        
    # Sobreescribir método string de la clase
    def __str__(self):
        return 'Nombre: {}\nNota: {}\n---\n'.format(self.nombre, self.nota)
    
    # Funcionalidades
    def calificacion(self):
        """ Método calificacion() de la clase Alumno. Evalúa la nota del alumno y muestra su calificación por pantalla.
        
            0 <  nota  < 5  - Suspenso
            5 <= nota  < 7  - Bien
            7 <= nota  < 9  - Notable
            9 <= nota  < 10 - Sobresaliente
            nota = 10       - Matrícula de honor
        """
        print(self.nombre, end='\t-\t')
        if self.nota < 5:
            print('Suspenso')
        else:
            print('Aprobado', end='\t-\t')
            if 5 <= self.nota < 7:
                print('¡Bien!')
            elif 7<= self.nota < 9:
                print('¡Notable!')
            elif 9<= self.nota < 10:
                print('¡Sobresaliente!')
            elif self.nota == 10:
                print('¡Matrícula de honor!')
            else:
                print('Error: No se pudo obtener su calificación', file=sys.stderr)
        
```

```python
# Clase Grupo
class Grupo():
    """ Clase que representa un grupo de alumnos """
    
    # Atributos de clase
    alumnos = list() # Lista de alumnos del catálogo
    
    # 
    def __init__(self, alumnos=[]):
        """ Constructor """
        self.alumnos = alumnos
        if len(self.alumnos) == 0:
            print('Se ha creado un grupo de alumnos vacío')
        else:
            print('Se ha creado un grupo de alumnos con contenido asignado')
            
    def agregar(self, a):
        """ Agregar un alumno al grupo
        @param a: alumno.
        """
        if isinstance(a, Alumno):
            self.alumnos.append(a) # Agregar alumno a la lista
            print('Se ha agregado al grupo el alumno: {}'.format(a.nombre))
        else:
            print('No se ha creado correctamente el alumno.', file=sys.stderr)
    
    def mostrar(self):
        """ Mostrar catálogo de películas """
        for a in self.alumnos:
            # Mostrar calificación del alumno
            a.calificacion()
            
    def numeroAlumnos(self):
        """ Devolver el número de alumnos"""
        return len(self.alumnos)
```

```python
# Experimentación (Crear diez alumnos)
# Nota. Se pasa directamente el objeto alumno para evitar almacenarlo en una variable.
# Crear grupo
grupo_python_uno = Grupo([Alumno('Sergio', 10), Alumno('Juan', 6.58), Alumno('Laura', 5.25), Alumno('Irene', 9.55)])
print('---')
```

```python
# Mostrar grupo
print('Grupo con',grupo_python_uno.numeroAlumnos(),'alumnos')
print('------')
grupo_python_uno.mostrar()
```

```python
# Añadir alumno
grupo_python_uno.agregar(Alumno('Daniel', 7.25))

# Mostrar grupo actualizado
print('Grupo con',grupo_python_uno.numeroAlumnos(),'alumnos')
print('------')
grupo_python_uno.mostrar()
```

## (2)  EJERCICIO 2.

### Creación
Sobre el ejercicio anterior realizaremos la siguiente modificación. Si en el ejercicio hacemos un __print(alumno1.nota)__, esto nos devolverá la nota del alumno. Esto puede llegar a ser un problma de seguridad. Vamos a encapsular el código para que no podamos acceder directamente a las variables, y el acceso tenga que ser a través de métodos.

* Haremos que las variables de la clase sean __privadas__. Recordatorio: Añadir doble guión delante del nombre de la variable.
* Crearemos métodos getter y setter para poder acceder y modificar las variables de la clase.

### Experimentación
* Crearemos algunos alumnos
* Comprobaremos que solo podamos acceder a las variables de la clase a través de sus método get y set.

```python
class Alumno():
    """ 
    Clase alumno.
        Atributos: - nombre. Nombre del alumno (str). Atributo privado.
                   - nota. Nota del alumno (float). Atributo privado.
    """
    # Atributos de la clase
    __nombre  = '' # Nombre del alumno 
    __nota = 0    # Nota del alumno
    
    # Constructor
    def __init__(self, nombre, nota):
        self.__nombre = nombre
        self.__nota = nota
        
        # Mostrar el objeto al crearlo.
        print(self, 'creado!')
    
    # Sobreescribir el método string
    def __str__(self):
        return 'Alumno: {}. Nota: {}'.format(self.__nombre, self.__nota)
    
    # Getter y Setter del atributo __nombre
    @property
    def nombre(self):
        """ Consultar atributo privado __nombre"""
        return self.__nombre
    
    # Setters
    @nombre.setter
    def nombre(self, nuevoValor):
        """ Asignar nuevo valor al atributo privado __nombre """
        self.__nombre = nuevoValor
        
    # Getter y Setter del atributo __nota
    @property
    def nota(self):
        """ Consultar atributo privado __nota """
        return self.__nota
    
    @nota.setter
    def nota(self, nuevoValor):
        """ Asignar nuevo valor al atributo privado __nota"""
        self.__nota = nuevoValor
        
```

```python
# Crear alumnos
al1 = Alumno('Sergio', 9.75)
al2 = Alumno('Irene', 7.25)
```

```python
class Producto:
    """ 
    Clase Producto con los atributos privados: codigo, nombre, precio, tipo.
    """
    # Atributos
    __codigo = None # Código de producto
    __nombre = None # Nombre de producto
    __precio = None # Precio de producto (€)
    __tipo   = None # Tipo de producto.
    
    # Constructor
    def __init__(self, codigo, nombre, precio, tipo):
        """ Constructor
        @param codigo: Codigo de producto
        @param nombre: Nombre del producto
        @param precio: Precio del producto
        @param tipo: Tipo del producto"""
        
        # Control de variables
        if isinstance(codigo, str):
            self.__codigo = codigo
        else:
            self.__codigo = None
            print('No se asignó el código al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
        if isinstance(nombre, str):
            self.__nombre = nombre
        else:
            self.__nombre = None
            print('No se asignó el nombre al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
            
        if (isinstance(precio, int) or isinstance(precio, float)):
            self.__precio = precio
        else:
            self.__precio = None
            print('No se asignó el precio al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
            
        if isinstance(tipo, str):
            self.__tipo = tipo
        else:
            self.__tipo = None
            print('No se asignó el tipo al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
        # Mensaje informando la creación.
        print('Se creará el siguiente producto:',end='\n\n')
        print(self)
        print('----')
            
    # Sobreescribir método str de python
    def __str__(self):
        return 'Código: {}\nNombre: {}\nPrecio: {}\nTipo: {}'.format(self.__codigo, self.__nombre, self.__precio, self.__tipo)
    
    # Getter y Setter del atributo privado __codigo
    @property
    def codigo(self):
        """ Getter del atributo privado __codigo """
        return self.__codigo
    @codigo.setter
    def codigo(self, nuevoValor):
        """ Setter del atributo privado __codigo """
        self.__codigo = nuevoValor
        
    # Getter y Setter del atributo privado __nombre
    @property
    def nombre(self):
        """ Getter del atributo privado __nombre """
        return self.__nombre
    @nombre.setter
    def nombre(self, nuevoValor):
        """ Setter del atributo privado __nombre """
        self.__nombre = nuevoValor
        
    # Getter y Setter del atributo privado __precio
    @property
    def precio(self):
        """ Getter del atributo privado __precio """
        return self.__precio
    @precio.setter
    def precio(self, nuevoValor):
        """ Setter del atributo privado __precio """
        self.__precio = nuevoValor
        
    # Getter y Setter del atributo privado __tipo
    @property
    def tipo(self):
        """ Getter del atributo privado __tipo """
        return self.__tipo
    @tipo.setter
    def tipo(self, nuevoValor):
        """ Setter del atributo privado __tipo """
        self.__tipo = nuevoValor
```

```python
import sys

SEPARADOR = 3 * '-'

class Pedido:
    """ Clase que define un pedido de productos """
    
    # Atributos
    __lista_productos = [] # Listado de productos del pedido
    __lista_cantidades = [] # Listado de cantidados asociadas a la __lista_productos
    
    # Constructor
    def __init__(self, lista_productos, lista_cantidades):
        """
        Constructor del pedido. 
            @param lista_productos: Listado de productos
            @param lista_cantidades: Listado de cantidades asociadasa a la lista_productos. 
        """
        
        # Asignación
        if isinstance(lista_productos, list):
            self.__lista_productos = lista_productos
        else:
            print('Error. La lista de productos no tiene el formato adecuado!', file=sys.stderr)
            
        
        if isinstance(lista_cantidades, list):
            self.__lista_cantidades = lista_cantidades
        else:
            print('Error. La lista de cantidades no tiene el formato adecuado!', file=sys.stderr)
            
        
        if len(lista_productos)!=len(lista_cantidades):
            print('Error. La lista de productos y cantidades no coinciden!', file=sys.stderr)
        else:
            print('¡Pedido creado con éxito!')
            
    def __calcular_pedido(self):
        """ Método privado para calcular el coste del pedido (encapsulación)"""
        precio_unitario = [producto.precio for producto in self.__lista_productos]

        # coste_total = sum(precio_unitario * cantidad_unidades). Almacenamos el coste a proposito en la clase para futuras implementaciones.
        coste_total = sum([p_unitario*cantidad for p_unitario,cantidad in zip(precio_unitario, self.__lista_cantidades)])
        return coste_total

    # Funcionalidades
    def total_pedido(self):
        """ Método total_pedido() de la clase Pedido. Devuelve el importe total de un pedido """
        # Devolver directamente el valor para evitar almacenar localmente el coste del pedido del cliente
        return self.__calcular_pedido()

    def __consultar_pedido(self):
        """ Método privado para mostrar el pedido realizado (encapsulación) """
        for a,b in zip(self.__lista_productos, self.__lista_cantidades):
            print(a, '\n\nCantidad:',b)
            print(SEPARADOR)
            
    def mostrar_pedido(self):
        """ Función mostrar_pedido() de la clase Pedido. Permite mostrar un pedido realizado """
        self.__consultar_pedido()
    

class Producto:
    """ 
    Clase Producto con los atributos privados: codigo, nombre, precio, tipo.
    """
    # Atributos
    __codigo = None # Código de producto
    __nombre = None # Nombre de producto
    __precio = None # Precio de producto (€)
    __tipo   = None # Tipo de producto.
        
    # Constructor
    def __init__(self, codigo, nombre, precio, tipo):
        """ Constructor
        @param codigo: Codigo de producto
        @param nombre: Nombre del producto
        @param precio: Precio del producto
        @param tipo: Tipo del producto"""
        
        # Control de variables
        if isinstance(codigo, str):
            self.__codigo = codigo
        else:
            self.__codigo = None
            print('No se asignó el código al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
        if isinstance(nombre, str):
            self.__nombre = nombre
        else:
            self.__nombre = None
            print('No se asignó el nombre al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
            
        if (isinstance(precio, int) or isinstance(precio, float)):
            self.__precio = precio
        else:
            self.__precio = None
            print('No se asignó el precio al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
        if isinstance(tipo, str):
            self.__tipo = tipo
        else:
            self.__tipo = None
            print('No se asignó el tipo al producto. Deberá hacerlo usted mismo', file=sys.stderr)
            
            
        # Mensaje informando la creación.
        print('Producto {} creado'.format(self.__codigo))
            
    # Funcionalidades
    def calcular_total(self, num_unidades):
        """ 
        Calcular precio total a partir de un número de unidades del producto.
        
            producto = num_unidades * precio
        
        """
        return round(num_unidades * self.precio, 2)
    
    # Getter y Setter del atributo privado __codigo
    @property
    def codigo(self):
        """ Getter del atributo privado __codigo """
        return self.__codigo
    @codigo.setter
    def codigo(self, nuevoValor):
        """ Setter del atributo privado __codigo """
        self.__codigo = nuevoValor
        
    # Getter y Setter del atributo privado __nombre
    @property
    def nombre(self):
        """ Getter del atributo privado __nombre """
        return self.__nombre
    @nombre.setter
    def nombre(self, nuevoValor):
        """ Setter del atributo privado __nombre """
        self.__nombre = nuevoValor
        
    # Getter y Setter del atributo privado __precio
    @property
    def precio(self):
        """ Getter del atributo privado __precio """
        return self.__precio
    @precio.setter
    def precio(self, nuevoValor):
        """ Setter del atributo privado __precio """
        self.__precio = nuevoValor
        
    # Getter y Setter del atributo privado __tipo
    @property
    def tipo(self):
        """ Getter del atributo privado __tipo """
        return self.__tipo
    @tipo.setter
    def tipo(self, nuevoValor):
        """ Setter del atributo privado __tipo """
        self.__tipo = nuevoValor
        
        
    # Sobreescribir método str de python
    def __str__(self):
        return 'Producto: {}\nNombre: {}\nPrecio: {}\nTipo: {}'.format(self.__codigo, self.__nombre, self.__precio, self.__tipo)
    
```

##### El plano cartesiano
Representa un espacio bidimensional en dos dimensiones, formado por dos rectas perpendiculares, una horizontal y otra vertical, que se cortan en un punto. 

La recta horizontal se denomina __eje de las abcisas__ o __eje X__, mientras que la vertical recibe el nombre de __eje de ordenadas__ o __eje Y__.

El punto donde se cortan las rectas, se conoce como __punto de origen (0, 0)__.

El plano se encuentra dividido en cuatro cuadrantes




##### Puntos y coordenadas.
El plano cartesiano permite describir la posición de puntos sobre el mismo en forma de coordenadas, que se forman asociando el valor del eje de las X (horizontal) con el valor del eje Y (vertical).

La representación de un punto es muy sencilla: __P(X,Y)__ dónde X, Y son la distancia horizontal (izquierda o derecha) y vertical (arriba o abajo) respectivamente, utilizando como referencia el punto de origen (0, 0), justo en el centro del plano.



##### Vectores en el plano
Un vector en el plano hace referencia a un segmento orientado, generado a partir de dos puntos distintos. A efectos prácticos, es una línea formada desde un punto inicial en dirección a otro punto final, por lo que se entiende que un vector tiene:
    * Longitud.
    * Dirección o sentido.
    


En esta figura, se observan dos puntos A, B definidos como:
* A(x1, y1) = A(2,3)
* B(x2, y2) = B(7,5)

Y el vector se representaría como la diferencia entre las coordenadas el egundo respecto al primero, es decir, la resta del segundo menos el primero.
* AB = (x2 - x1) / (y2 - y1)

Lo que, en definitiva no deja de ser:
* Cinco unidades a la derecha
* Dos unidades hacia arriba

```python
import sys
import math
import operator

SEPARADOR = 5 * '-'

class Punto:
    """ Clase que define un punto en el plano cartesiano """
    # Atributos
    __x = 0 # Valor de la coordenada X (privado).
    __y = 0 # Valor de la coordenada Y (privado).
    
    # Constructor
    def __init__(self, x=0, y=0):
        """ Constructor de una instancia de la clase Punto. Valor por defecto (0,0) """
        # Asignación.
        self.__x = x
        self.__y = y
    
    # Getter
    @property
    def x(self):
        """ Getter atributo privado __x"""
        return self.__x
    @property
    def y(self):
        """ Getter atributo privado __y"""
        return self.__y
    
    # Funcionalidades
    def __obtener_cuadrante(self):
        """ Método __obtener_cuadrante() de la clase Punto (método privado). """
        if self.__x > 0 and self.__y > 0:
            return 'Primer cuadrante (+, +). [0, 𝞹/2]'
        elif self.__x < 0 and self.__y > 0:
            return 'Segundo cuadrante (-, +) [𝞹/2, 𝞹]'
        elif self.__x < 0 and self.__y < 0:
            return 'Tercer cuadrante (-,-). [𝞹, 3𝞹/2]'
        elif self.__x > 0 and self.__y < 0:
            return 'Cuarto cuadrante (+, -). [3𝞹/2, 0]'
        elif self.__x == 0 and self.__y == 0:
            return 'Origen de coordenadas (0,0)'
        else:
            return 'El punto se encuentra sobre un eje'
        
    def cuadrante(self):
        """ Método cuadrante() de la clase Punto. Devuelve el cuadrante al cuál pertenece el punto """
        return self.__obtener_cuadrante()
    
    def vector(self,p_destino):
        """ Método vector() de la clase Punto. Calcula el vector tomando el punto como origen y el argumento de entrada
        como destino """
        return (p_destino.x - self.x, p_destino.y - self.y)
    
    def distancia(self, p_destino):
        """ Método destino() de la clase Punto. Calcula la distancia tomando el punto como origen y el argumento de entrada
        como destino """
        vector = self.vector(p_destino)
        return math.sqrt(pow(vector[0],2) + pow(vector[1],2))  
        
    # Soobrescribir método string
    def __str__(self):
        return '({}, {})'.format(self.__x, self.__y)
```

```python
class Rectangulo:
    """ Clase rectángulo que se crea utilizando dos puntos (inicial y final, atributos de la clase) que formarán la diagonal
        del triángulo. """
    
    # Atributos
    __inicial = None   # Punto inicial de la diagonal rectángulo.
    __final   = None   # Punto final de la diagonal del rectángulo.
    
    
    # Variables de clase
    __diagonal = None # Diagonal de la diagonal.
    __vector   = None # Vector que define la diagnoal
    
    # Constructor
    def __init__(self, inicial=Punto(0, 0), final=Punto(0, 0)):
        """ Constructor de un rectángulo"""
        self.__inicial   = inicial
        self.__final     = final
        self.vector      = self.final.x - self.inicial.x, self.final.y - self.inicial.y # Vector que define la diagonal
        self.diagonal    = math.sqrt(pow(self.vector[0],2) + pow(self.vector[1],1))     # distancia (módulo del vector que define la diagonal)
        
    # Getter.
    @property
    def inicial(self):
        """ Getter de la variable privada inicial."""
        return self.__inicial
    
    @property 
    def final(self):
        """ Getter de la variable privada final. """
        return self.__final
    
    @property
    def diagonal(self):
        """ Getter de la variable privada distancia. """
        return self.__diagonal
    
    @property
    def vector(self):
        """ Getter de la variable privada vector. """
        return self.__vector
    
    # Setter
    @diagonal.setter
    def diagonal(self, nuevoValor):
        """ Setter de la variable privada diagonal"""
        self.__diagonal = nuevoValor  
    
    @vector.setter
    def vector(self, nuevoValor):
        """ Setter de la variable privada vector """
        self.__vector = nuevoValor
    
    # Funcionalidades
    def base(self):
        """ Calcular base del rectángulo """
        return abs(self.vector[0])
    
    def altura(self):
        """ Calcular la altura del rectángulo """
        return abs(self.vector[1])
    
    def area(self):
        """ Calcular el área del rectángulo """
        return abs(self.vector[0] * self.vector[1])
    
    # Sobreescribir el método string
    def __str__(self):
        return '\tVertice Inferior Izquierdo: {}\n\tVertice Superior Derecho: {}'.format(self.__inicial, self.__final)
```

```python
import sys

SEPARADOR = 5*'-'
class Vehiculo():
    """ 
    Clase que define a un Vehiculo (superclase) con los siguientes atributos:
        - color
        - ruedas
    """
    
    # Atributos
    __color  = '' # Color del vehículo
    __ruedas = '' # Número de ruedas del vehículo
    
    # Constructor
    def __init__(self, color, ruedas):
        """ Constructor de la clase Vehiculo """
        # Asignación
        self.__color  = color
        self.__ruedas = ruedas
    
    # Getters
    @property
    def color(self):
        return self.__color
    @property
    def ruedas(self):
        return self.__ruedas
    
    # Sobreescribir función string
    def __str__(self):
        return 'Información del vehículo:\n\tColor: {}\n\tRuedas: {}'.format(self.color, self.ruedas)
    
class Coche(Vehiculo):
    """ 
    Clase que define un Coche y que hereda de Vehiculo con los siguientes atributos:
        - velocidad (km/h)
        - cilindrada (cm3)
    """
    
    # Atributos
    velocidad = 0 # Velocidad del vehículo (km/h)
    __cilindrada = 0 # Cilindrada del vehículo (cm3 o centímetros cúbicos)
    
    # Constructor
    def __init__(self, color, ruedas, velocidad, cilindrada):
        """ Constructor de la clase Coche que hereda de la clase Vehiculo"""
        super().__init__(color, ruedas)
        self.velocidad = velocidad
        self.__cilindrada = cilindrada
        
    # Getter y setter del atributo privado cilindrada
    @property
    def cilindrada(self):
        """ Getter del atributo privado cilindrada """
        return self.__cilindrada
    
    @cilindrada.setter
    def cilindrada(self, nuevoValor):
        """ Setter del atributo privado cilindrada """
        self.__cilindrada = nuevoValor
        
    # Sobreescribir método string
    def __str__(self):
        return '{}\nInformación del coche:\n\tVelocidad: {}\n\tCilindrada: {}'.format(super().__str__(),self.velocidad, self.cilindrada)
```

```python
import sys

SEPARADOR = 115*'-'
class Vehiculo():
    """ 
    Clase que define a un Vehiculo (superclase) con los siguientes atributos:
        - color
        - ruedas
    """
    
    # Atributos
    __color  = '' # Color del vehículo
    __ruedas = '' # Número de ruedas del vehículo
    
    # Constructor
    def __init__(self, color, ruedas):
        """ Constructor de la clase Vehiculo """
        # Asignación
        self.__color  = color
        self.__ruedas = ruedas
    
    # Getters
    @property
    def color(self):
        return self.__color
    @property
    def ruedas(self):
        return self.__ruedas
    
    # Sobreescribir función string
    def __str__(self):
        return 'Color: {} - Ruedas: {}'.format(self.color, self.ruedas)
```

```python
    
class Coche(Vehiculo):
    """ 
    Clase que define un Coche y que hereda de Vehiculo con los siguientes atributos:
        - velocidad (km/h)
        - cilindrada (cm3)
    """
    
    # Atributos
    velocidad = 0 # Velocidad del vehículo (km/h)
    __cilindrada = 0 # Cilindrada del vehículo (cm3 o centímetros cúbicos)
    
    # Constructor
    def __init__(self, color, ruedas, velocidad, cilindrada):
        """ Constructor de la clase Coche que hereda de la clase Vehiculo"""
        super().__init__(color, ruedas)
        self.velocidad = velocidad
        self.__cilindrada = cilindrada
        
    # Getter y setter del atributo privado cilindrada
    @property
    def cilindrada(self):
        """ Getter del atributo privado cilindrada """
        return self.__cilindrada
    
    @cilindrada.setter
    def cilindrada(self, nuevoValor):
        """ Setter del atributo privado cilindrada """
        self.__cilindrada = nuevoValor
        
    # Sobreescribir método string
    def __str__(self):
        return '{} - Velocidad: {} - Cilindrada: {}'.format(super().__str__(), self.velocidad, self.cilindrada)
```

```python
class Bicicleta(Vehiculo):
    """ Clase Bicicleta que hereda de la clase Vehiculo """
    # Atributos
    tipo = 'urbana'
    
    # Variables de clase
    __categorias = ['urbana', 'deportiva']
    
    # Constructor
    def __init__(self, color, ruedas, tipo=None):
        """ Constructor de la clase Bicicleta que recibe como argumento de entrada el tipo (dentro de las posibles categorias) """
        super().__init__(color, ruedas)
        if tipo in self.__categorias and tipo is not None:
            self.tipo = tipo
        else:
            # Por defecto, si no se introduce una categoría correcta o no se introduce ninguna categoría, se selecciona urbana.
            pass
    
    # Sobreescribir método string
    def __str__(self):
        return '{} - Tipo: {}'.format(super().__str__(), self.tipo)
```

```python
class Motocicleta(Bicicleta):
    """ Clase Motocicleta que hereda de la clase Bicicleta"""
    # Atributos
    velocidad = 0 # Velocidad de la bicicleta por defecto
    __cilindrada = 0 # Cilindrada del vehículo por defecto
    
    # Constructor
    def __init__(self, color, ruedas, tipo, velocidad, cilindrada):
        """ Constructor """
        super().__init__(color, ruedas, tipo)
        self.velocidad = velocidad
        self.__cilindrada = cilindrada
        
    # Getters
    @property
    def cilindrada(self):
        """ Getter del atributo privado cilindrada """
        return self.__cilindrada
    
    # Sobreescribir método string
    def __str__(self):
        return '{} - Velocidad: {} - Cilindrada: {}'.format(super().__str__(), self.velocidad, self.cilindrada)
```
