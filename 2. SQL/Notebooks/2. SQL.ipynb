### Lista de oficinas con la cantidad vendida por vendedor (Considerar todas las oficinas)
Tip: a agrupar toca

```python
# Lista de oficinas que no tienen vendedores asignados
sSQL = """
    SELECT o.ID_OFICINA, o.DESCRIPCION AS oficina, v.ID_VENDEDOR, v.NOMBRE, v.APELLIDO, COUNT(vs.ID_VENDEDOR) AS CANTIDAD_VENDIDA
    FROM vendedores AS v
    RIGHT JOIN oficinas AS o
    ON v.ID_OFICINA = o.ID_OFICINA
    LEFT JOIN ventas AS vs
    ON v.ID_VENDEDOR = vs.ID_VENDEDOR
    GROUP BY o.ID_OFICINA, o.DESCRIPCION, v.ID_VENDEDOR, v.NOMBRE, v.APELLIDO
    ORDER BY o.ID_OFICINA, v.ID_VENDEDOR
"""
pd.read_sql(sSQL, con=con)
```

```python
# Lista de oficinas con la cantidad vendida por vendedor (Considerar todas las oficinas) que hayan vendido más de 10 unidades
sSQL = """
    SELECT o.ID_OFICINA, o.DESCRIPCION AS oficina, v.ID_VENDEDOR, v.NOMBRE, v.APELLIDO, COUNT(vs.ID_VENDEDOR) AS CANTIDAD_VENDIDA
    FROM vendedores AS v
    RIGHT JOIN oficinas AS o
    ON v.ID_OFICINA = o.ID_OFICINA
    LEFT JOIN ventas AS vs
    ON v.ID_VENDEDOR = vs.ID_VENDEDOR
    GROUP BY o.ID_OFICINA, o.DESCRIPCION, v.ID_VENDEDOR, v.NOMBRE, v.APELLIDO
    HAVING CANTIDAD_VENDIDA > 10
    ORDER BY o.ID_OFICINA, v.ID_VENDEDOR
"""
pd.read_sql(sSQL, con=con)
```

### Ingresos de ventas entre el 15 de enero y el 15 de febrero (los precios están en EUROS)
Tip: revisad cómo están formateadas las fechas para poder hacer comparaciones

```python
# Ingresos de ventas entre el 15 de enero al 15 de febrero
sSQL = """
    SELECT SUM(p.PRECIO) AS INGRESOS_EUR, SUM(p.PRECIO)*1.12 AS INGRESOS_USD 
    FROM ventas AS v
    LEFT JOIN productos AS p
    ON p.ID_PRODUCTO = v.ID_PRODUCTO
    WHERE v.FECHA BETWEEN '2020-01-15' AND '2020-02-15'
"""
pd.read_sql(sSQL, con=con)
```

### Lista de productos vendidos (Con su descripción, precio, cantidad vendidas e ingresos) entre el 15 de enero y el 15 de febrero (los precios están en EUROS)
Tip: para los que habéis hecho economía, ingresos = p*q (precio por cantidad :-) )

```python
# Lista de productos vendidos (Con su descripción, precio, cantidad vendidas e ingresos) --> CON PRODUCTO CARTESIANO
sSQL = """
    SELECT p.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO, COUNT(v.ID_VENTA) AS CANTIDAD_VENDIDA, COUNT(V.ID_VENTA)*p.PRECIO AS INGRESOS
    FROM ventas as v, 
    productos AS p
    WHERE 
        v.ID_PRODUCTO = p.ID_PRODUCTO
        AND v.FECHA BETWEEN  '2020-01-15' AND '2020-02-15'
    GROUP BY v.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO
"""
pd.read_sql(sSQL, con=con)
```

```python
# Lista de productos vendidos (Con su descripción, precio, cantidad vendidas e ingresos) --> CON UN LEFT JOIN 
sSQL = """
    SELECT p.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO, COUNT(v.ID_VENTA) AS CANTIDAD_VENDIDA, COUNT(V.ID_VENTA)*p.PRECIO AS INGRESOS
    FROM ventas as v 
    INNER JOIN productos AS p
    ON v.ID_PRODUCTO = p.ID_PRODUCTO
    WHERE 
        v.FECHA BETWEEN  '2020-01-15' AND '2020-02-15'
    GROUP BY v.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO
"""
pd.read_sql(sSQL, con=con)
```

```python
# Lista de productos vendidos (Con su descripción, precio, cantidad vendidas e ingresos)
# VAMOS A UNIR LOS DE FECHA ANTERIOR AL 15/01 CON LOS DE FECHA ENTRE 15/01 Y EL 15/02
# UTILIZAMOS UN UNION

sSQL = """
    
    SELECT CASE WHEN v.FECHA < '2020-01-15' THEN 'ANTES DEL 15/01' ELSE 'DESPUES 15/01' END AS RANGO_FECHA, p.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO, COUNT(v.ID_VENTA) AS CANTIDAD_VENDIDA, COUNT(V.ID_VENTA)*p.PRECIO AS INGRESOS
    FROM ventas as v 
    INNER JOIN productos AS p
    ON v.ID_PRODUCTO = p.ID_PRODUCTO
    WHERE 
        v.FECHA <  '2020-01-15'
    GROUP BY v.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO, v.FECHA

    UNION

    SELECT CASE WHEN v.FECHA < '2020-01-15' THEN 'ANTES DEL 15/01' ELSE 'DESPUES 15/01' END AS RANGO_FECHA, p.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO, COUNT(v.ID_VENTA) AS CANTIDAD_VENDIDA, COUNT(V.ID_VENTA)*p.PRECIO AS INGRESOS
    FROM ventas as v 
    INNER JOIN productos AS p
    ON v.ID_PRODUCTO = p.ID_PRODUCTO
    WHERE 
        v.FECHA BETWEEN  '2020-01-15' AND '2020-02-15'
    GROUP BY v.ID_PRODUCTO, p.DESCRIPCION, p.PRECIO, v.FECHA

"""
pd.read_sql(sSQL, con=con)
```

### Ranking de mejores vendedores (criterio: por cantidad de productos vendidos)

```python
# Ranking de mejores vendedores (criterio: por cantidad de productos vendidos)
sSQL = """
    SELECT ve.ID_VENDEDOR, ve.NOMBRE, ve.APELLIDO, COUNT(v.ID_PRODUCTO) AS CANTIDAD_VENDIDA
    FROM vendedores AS ve
    INNER JOIN ventas AS v
    ON ve.ID_VENDEDOR = v.ID_VENDEDOR 
    GROUP BY ve.ID_VENDEDOR, ve.NOMBRE, ve.APELLIDO
    ORDER BY CANTIDAD_VENDIDA DESC
    LIMIT 0, 3
"""
pd.read_sql(sSQL, con=con)
```

```python
# Ranking de los peores vendedores (criterio: por cantidad de productos vendidos)
sSQL = """
    SELECT ve.ID_VENDEDOR, ve.NOMBRE, ve.APELLIDO, COUNT(v.ID_PRODUCTO) AS CANTIDAD_VENDIDA
    FROM vendedores AS ve
    LEFT JOIN ventas AS v
    ON ve.ID_VENDEDOR = v.ID_VENDEDOR 
    GROUP BY ve.ID_VENDEDOR, ve.NOMBRE, ve.APELLIDO
    ORDER BY CANTIDAD_VENDIDA ASC
"""
pd.read_sql(sSQL, con=con)
```

### Ranking de mejores clientes (criterio: por ingresos que obtiene la empresa EN EUROS)

```python
# Ranking de mejores clientes (criterio: por ingresos que obtiene la empresa)
sSQL = """
    SELECT CONCAT(c.NOMBRE, ' ', c.APELLIDO) AS CLIENTE, SUM(p.PRECIO) AS INGRESOS_EUR
    FROM clientes c
    INNER JOIN ventas as v
    ON c.ID_CLIENTE = v.ID_CLIENTE
    INNER JOIN productos as p
    ON p.ID_PRODUCTO = v.ID_PRODUCTO
    GROUP BY v.ID_CLIENTE, c.NOMBRE, c.APELLIDO
    ORDER BY INGRESOS_EUR DESC
"""
pd.read_sql(sSQL, con=con)
```

### Ranking de mejores clientes que tengan como vendedor la oficina de Barcelona (criterio: por ingresos que obtiene la empresa)

```python
# Ranking de mejores clientes que tengan como vendedor la oficina de Barcelona (criterio: por ingresos que obtiene la empresa)
sSQL = """
    SELECT CONCAT(cts.NOMBRE , ' ', cts.APELLIDO) AS CLIENTE, o.DESCRIPCION AS OFICINA, sum(pds.PRECIO) AS INGRESOS
    FROM ventas AS vs
    INNER JOIN clientes AS cts
    ON vs.ID_CLIENTE = cts.ID_CLIENTE
    INNER JOIN vendedores AS v
    ON v.ID_VENDEDOR = vs.ID_VENDEDOR
    INNER JOIN oficinas as o
    ON o.ID_OFICINA = v.ID_OFICINA
    INNER JOIN productos as pds
    ON pds.ID_PRODUCTO = vs.ID_PRODUCTO
    WHERE o.DESCRIPCION = 'BARCELONA' 
    GROUP BY vs.ID_CLIENTE, cts.NOMBRE, cts.APELLIDO, o.DESCRIPCION
    ORDER BY INGRESOS DESC;
"""
pd.read_sql(sSQL, con=con)
```

### Conteo de productos que se hayan vendido más de 10 veces

```python
# Conteo de productos que se hayan vendido más de 10 veces
sSQL = """
    SELECT p.ID_PRODUCTO, p.DESCRIPCION, COUNT(v.ID_PRODUCTO) AS PRODUCTOS_VENDIDOS
    FROM productos as p
    INNER JOIN ventas AS v
    ON p.ID_PRODUCTO = v.ID_PRODUCTO
    GROUP BY p.ID_PRODUCTO, p.DESCRIPCION
    HAVING PRODUCTOS_VENDIDOS > 10
    ORDER BY PRODUCTOS_VENDIDOS DESC
"""
pd.read_sql(sSQL, con=con)
```

### Un reporte final de ventas con el siguiente formato
### Nombre y apellido del vendedor/oficina del vendedor/nombre y apellido del cliente/nombre del producto y precio/fecha de venta

```python
# Nombre y apellido del vendedor/oficina del vendedor/nombre y apellido del cliente/nombre del producto y precio/fecha de venta
sSQL = """
    SELECT o.DESCRIPCION, CONCAT(v.NOMBRE, ' ', v.APELLIDO) as VENDEDOR, vts.FECHA, CONCAT(c.NOMBRE, ' ', c.APELLIDO) AS CLIENTE, pds.DESCRIPCION as PRODUCTO, pds.PRECIO 
    FROM vendedores AS v
    RIGHT JOIN oficinas as o
    ON o.ID_OFICINA = v.ID_OFICINA
    INNER JOIN ventas as vts
    ON vts.ID_VENDEDOR = v.ID_VENDEDOR
    INNER JOIN productos AS pds
    ON pds.ID_PRODUCTO = vts.ID_PRODUCTO
    INNER JOIN clientes as c
    ON c.ID_CLIENTE = vts.ID_CLIENTE
    ORDER BY vts.FECHA ASC
"""
pd.read_sql(sSQL, con=con)
```

```python
# Nombre y apellido del vendedor/oficina del vendedor/nombre y apellido del cliente/nombre del producto y precio/fecha de venta
sSQL = """
    SELECT tabla.CLIENTE, SUM(PRECIO)
    FROM (
        SELECT o.DESCRIPCION, CONCAT(v.NOMBRE, ' ', v.APELLIDO) as VENDEDOR, vts.FECHA, CONCAT(c.NOMBRE, ' ', c.APELLIDO) AS CLIENTE, pds.DESCRIPCION as PRODUCTO, pds.PRECIO 
        FROM vendedores AS v
        RIGHT JOIN oficinas as o
        ON o.ID_OFICINA = v.ID_OFICINA
        INNER JOIN ventas as vts
        ON vts.ID_VENDEDOR = v.ID_VENDEDOR
        INNER JOIN productos AS pds
        ON pds.ID_PRODUCTO = vts.ID_PRODUCTO
        INNER JOIN clientes as c
        ON c.ID_CLIENTE = vts.ID_CLIENTE
        ORDER BY vts.FECHA ASC
    ) AS tabla
    GROUP BY tabla.CLIENTE
"""
pd.read_sql(sSQL, con=con)
```
