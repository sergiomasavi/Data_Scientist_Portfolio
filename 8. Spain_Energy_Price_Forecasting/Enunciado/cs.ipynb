# Previsión del precio energético en España

La producción y demanda de energía es, sin duda, uno de los factores más importantes del progreso de la sociedad. La electricidad ha entrado en el mercado como un producto comercializable y el sector eléctrico de muchos países se ha desregulado. En España, la Ley 54/1997 de la Energía Eléctrica expuso a todas las partes interesadas a una gran incertidumbre, ya que:

- Su precio viene determinado por innumerables factores.

- No puede almacenarse en grandes cantidades de forma eficiente. 

Esta complejidad ha hecho que surja la necesidad de contar con métodos de previsión fiables a todas las escalas (horaria, diaria, mensual, anual, etc.) y se ha convertido en un amplio campo de investigación.

En este trabajo el objetivo es comparar diferentes técnicas de Machine Learning aplicadas a series temporales con el objetivo de predecir el precio diario de la electricidad utilizando los valores pasados del precio de la electricidad así como otras características relacionadas con la generación de energía y las condiciones meteorológicas. Además, el este coumente contiene la exploración y limpieza de los datos, un análisis de series temporales del precio de la electricidad e ingeniería de características. 

En el proyecto original, realizado para un curso de posgrado, comparé el rendimiento (utilizando el error medio cuadrático como métrica de rendimiento) de 5 arquitecturas diferentes (LSTM, LSTM apilado, CNN, CNN-LSTM y MLP distribuido en el tiempo) tanto para la previsión univariante como multivariante (es decir, utilizando sólo los pasos temporales anteriores). es decir, utilizando sólo los pasos de tiempo anteriores del precio de la electricidad frente a la utilización de otras características) utilizando un número diferente de pasos de tiempo anteriores como características para los modelos (3, 10 y 25 pasos de tiempo anteriores para todas las características utilizadas). En este núcleo en particular, encontrarás una aplicación de todas las arquitecturas de aprendizaje profundo mencionadas anteriormente, así como dos enfoques más: la arquitectura Encoder-Decoder y el regresor XGBoost. Además, en todas estas aplicaciones, utilizo los 25 pasos de tiempo anteriores de todas las características (predicción multivariante) que se han extraído o generado, tras aplicar el PCA (Análisis de Componentes Principales). El optimizador Adam se utiliza en todas las arquitecturas de aprendizaje profundo y, para elegir su tasa de aprendizaje, originalmente llevé a cabo pruebas preliminares utilizando la devolución de llamada del programador de la tasa de aprendizaje, empezando por una tasa de aprendizaje igual a 0,0001 y aumentándola gradualmente en un factor de 10 cada 10 épocas; un paso que se omite en este núcleo.

Los estadísticos entre ambas series temporales no son iguales pero si son muy similares. Se visualizan ambas para comprobar su parecido.

```python
def plot_lines(df, lines, title_text, ylabel_text):

    # Gráfico evolución temporal del precio de la energía.
    fig = go.Figure()
    for col_name in lines:
        fig.add_trace(
            go.Scatter(
                x=df['date'], 
                y=df[col_name], 
                mode='lines', 
                name=col_name,
                )

        )

    fig.update_layout(title_text=title_text)
    fig.update_yaxes(title_text=ylabel_text)
    fig.update_traces(hovertemplate="%{y:.2f}€")

    fig.update_yaxes(tickprefix="")

    fig.update_xaxes(
        showspikes=True, spikethickness=2, spikecolor="#999999", spikemode="across"
    )

    fig.update_layout(
        hovermode="x unified",
        hoverdistance=200,
        spikedistance=200,
        transition_duration=500,
    )
    fig.show()
```

Dado que la serie horaria es muy ruidosa y no permite apreciar la visualización, se pasa a una serie mensual para la comprobación

```python
plot_lines(df=df_check.resample('m').mean().reset_index(), 
           lines=['ElectroStock', 'RedElectrica'],
           title_text='Evolución temporal del precio promedio mensual energético en España',
           ylabel_text='Precio MWh (€)')
```

```python
def is_normal(r, level=0.01):
    """
    Aplica el test de Jarque-Bera para determinar si una Serie es normal o no
    
    
    Args:
    ------
    r {[pandas.Series]} -- Serie temporal de rendimientos
    level {[float]} -- Nivel de confianza del test. La prueba se aplica por defecto al nivel del 1%


    Returns:
    ---------
    aceptar_hipotesis {[bool]} -- Devuelve Verdadero si se acepta la hipótesis de normalidad, Falso en caso contrario
    """
    if isinstance(r, pd.DataFrame):
        return r.aggregate(is_normal)
    else:
        statistic, p_value = scipy.stats.jarque_bera(r)
        return (p_value > level), p_value, level
```

```python
resultado_test, p_value, level = is_normal(r=dataset['AdjPrice'], level=0.01)
skewness = scipy.stats.skew(dataset['AdjPrice'])
kurtosis = scipy.stats.kurtosis(dataset['AdjPrice'])
mean = dataset['AdjPrice'].mean()
median = dataset['AdjPrice'].median()
mean_median = mean > median

print('Test normalidad distribución retornos:')
pd.Series({
    'normal': resultado_test, 
    'p_value': p_value,
    'level':level,
    'skewness_excess': skewness,
    'kurtosis': kurtosis,
    'mean':mean,
    'median':median,
    'mean>median': mean_median
}).to_frame('Test Jarque Bera')
```

### Estacionalidad y descomposición temporal


Los datos de series de tiempo pueden exhibir una variedad de patrones y, por lo tanto, es útil dividir una serie temporal en varios componentes, cada uno de los cuales representa una categoría de patrón subyacente.

las componentes principales de una serie temporal son:

- Tendencia. Existe una tendencia cuando hay un aumento o disminución a largo plazo en los datos (lineal o no lineal) y también se conoce como "cambio de dirección", cuando puede pasar de una tendencia creciente a una tendencia decreciente. 
    
    
- Estacional. Un patrón estacional ocurre cuando una serie temporal se ve afectada por factores estacionales como la época del año o el día de la semana. La estacionalidad es siempre de una frecuencia fija y conocida. 


- Cíclico. Un ciclo ocurre cuando la exhibición de datos sube y baja sin una frecuencia fija. Estas fluctuaciones generalmente se deben a las condiciones económicas y, a menudo, están relacionadas con el "ciclo económico". La duración de estas fluctuaciones suele ser de al menos 2 años. Es importante no confundir el comportamiento cíclico con el comportamiento estacional. Si las fluctuaciones no son de una frecuencia fija, entonces son cíclicas; si la frecuencia no cambia y está asociada con algún aspecto del calendario, entonces el patrón es estacional. En general, la duración media de los ciclos es mayor que la duración de un patrón estacional, y las magnitudes de los ciclos tienden a ser más variables que las magnitudes de los patrones estacionales.


- Componente de residuos. Contiene cualquier otra información en la serie.

Al elegir un método de pronóstico, primero es necesario identificar los patrones de la serie temporal en los datos y luego elegir un método que sea capaz de capturar los patrones correctamente.

La noción de estacionariedad es una abstracción útil a la hora de pronosticar.

- Proceso estacionario. Proceso que genera una serie estacionaria de observaciones.


- Modelo estacionario. Modelo que describe una serie estacionaria de observaciones.


- Tendencia estacionaria. Serie de tiempo que no muestra una tendencia.


- Estacional estacional. Serie de tiempo que no exhibe estacionalidad.


- Estrictamente estacionario. Definición matemática de un proceso estacionario, específicamente que la distribución conjunta de observaciones es invariante al cambio de tiempo.

Las pruebas estadísticas hacen suposiciones sólidas sobre sus datos y se utilizan para informar el grado en que una hipótesis nula puede rechazarse o dejar de ser rechazada. Por lo tanto, se debe interpretar el resultado de los test estadísticas para dar explicabilidad.

Sin embargo, pueden 

La prueba Augmented Dickey-Fuller es un tipo de prueba estadística llamada prueba de raíz unitaria que proporciona una verificación rápida y evidencia confirmatoria de si una serie temporal es estacionaria o no estacionaria. La intuición detrás de una prueba de raíz unitaria es que determina la fuerza con la que una serie de tiempo está definida por una tendencia.

El test Augmented Dickey-Fuller es muy utilizado como prueba de raíz unitaria. Utiliza un modelo autorregresivo y optimiza un criterio de información a través de múltiples valores de retardo diferentes. Las hipotesis que utiliza son:

+ **Hipótesis nula (H0)**: Sugiere que la serie temporal tiene raíz unitaria, lo que significa que no es estacionaria, es decir, tiene una estructura dependiente del tiempo.


+ **Hipótesis alternativa (H1**): Si se rechaza la hipótesis nula, sugiere que la serie temporal no tiene una raíz unitaria, lo que significa que es estacionaria, es decir, no tiene una estructura dependiente del tiempo.

Interpretamos este resultado utilizando el p-value del test. Cuando se sitúa por debajo de un umbral (como el 5% o el 1%) indica que se rechaza la hipótesis nula, es decir, es estacionaria. Sin embargo, cuando se encuentra por encima del umbral indica que no se rechaza la hipótesis nula, es decir, no es estacionaria.

+ **Valor p > threshold**: No se rechaza la hipótesis nula (H0), los datos tienen raíz unitaria y NO son estacionarios, es decir, tiene una estructura dependiente del tiempo.


+ **Valor p <= threshold**: Rechaza la hipótesis nula (H0), los datos no tienen una raíz unitaria y son estacionarios, es decir, no tiene estructura dependiente del tiempo

### SARIMAX

Los modelos ARIMA está diseñado para el modelado de series temporales teniendo en cuenta la dependencia entre los
datos, es decir cada observación en un momento dado, se calcula en función de los anteriores.El valor final es una función lineal de datos pasados y de los errores debidos al azar. Además, en sus variantes, permite añadir información estacional y variables exteriores para completar el modelo. Para la resolución, existe la metodología de Box y Jenkins.

La metodologia de Jen Bookings nos permite resolver los problemas de series temporales por pasos:

- Identificación. Se utiliza la serie temporal “Y” así como cualquier información adicional disponible para “acotar “ un conjunto de posibles modelos que hayan podido generar la serie temporal. Las herramientas que se utilizan son gráficos como la autocorrelación y/o la autocorrelación parcial para elegir el grado del modelo ARMA (p,q).


- Selección. En esta fase se estiman utilizando software informático los coeficientes del modelo que hemos “acotado” en el primer paso.


- Diagnóstico. Se realizan test estadísticos para ver la bondad de los modelos. Normalmente se analiza la distribución de los residuos (si siguen una distribución normal).


- Predicción. Una vez que tenemos el mejor modelo, podemos utilizarlo para hacer inferencia a futuro (forecasting).

### PROPHET


La biblioteca Prophet es una librería open-source desarrollada por Facebook y diseñada para la previsión automática de datos de series temporales. Este modelo se centra principalmente en la predicción de tendencias y estacionalidad basado en un modelo aditivio en el que las tendencias no lineales se ajustan a la estacionalidad anual, semanal y diaria. El modelo funciona mejor en series temporales que presentan fuertes efectos estacionales (semanales, mensuales) y una cantidad suficiente de datos históricios donde queda representada dicha estacionalidad. Es robusto a los missing data y a los cambio sde tendencia y, además, suele manejar bien los outliers. 

El uso de Prophet se basa en la API sklearn que consiste en crear una instancia de la clase Prophet y luego llamar a sus métodos de ajuste y predicción.

La entrada a Prophet es siempre un marco de datos con dos columnas: 

- ds (datestamp). Esta columna debe tener el formato esperado por Pandas, idealmente YYYY-MM-DD para una fecha o YYYY-MM-DD HH:MM:SS para una marca de tiempo. 

- y. Debe ser numérica, y representa la medida que se desea pronosticar.


Factores principales para operar con Prophet.

- Se puede incluir información hora, diaria, y se necesita una muestra de algunos meses (un año si se puede incluir).


- Se tienen en cuenta temporalidades que afectan: dia de la semana, estacion del año, y eventos.


- Puede manejar NAs y outliers.


- Tambien tiene en cuenta tendencias históricas y eventos concretos que afecten a la variable target.


- Del mismo modo, es posible detectar tendencias de evolución que no son lineales.

```python
def convert(tude):
    prev = 0
    multiplier = 1 if tude[-1] in ['N', 'E'] else -1

    cord=list()
    for i in range(1,4):
        lim = 2*i
        if prev==0: 
            cord.append(int(tude[:lim]))
        else:
            cord.append(int(tude[prev:lim]))
        prev = lim

    decimal = multiplier*(cord[0] + cord[1] / 60 + cord[2]/3600)

    return decimal
```

```python
def crear_map(df):
    folium_map = folium.Map(location=[40,-3], tiles="cartodbpositron", zoom_start=6.2)
    for i, row in df.iterrows():
        folium.Circle(
            location=[row['Lat'], row['Long']], 
            color="crimson", 
            radius=100,
            tooltip="""
            <uL>
                <li>ID: {}</li>
                <li>Codigo: {}</li>
                <li>Municipio: {}</li>
                <li>Provinvicia: {}</li>
                <li>Inicio: {}</li>
                <li>Fin: {}</li>
            </ul>
            """.format(row["ID"], row["Codigo"], row["Municipio"], row["Provincia"],
                       row['Inicio'], row['Fin'])
        ).add_to(folium_map)
        
        
        
    return folium_map
```

```python
def read_data(ruta_base):
    data = list()
    arr = os.listdir(ruta_base)
    for filename in tqdm(arr):
        df = pd.read_csv(ruta_base+filename, sep=';')
        data.append(df)
            
    df_data = pd.concat(data, axis=0)
    return df_data
```
