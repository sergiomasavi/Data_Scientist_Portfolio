## Descomposicion de Series Temporales

La descomposicin de series temporales nos ayuda a comprender el valor de una serie temporal, y sus componentes ms importantes. Se puede hacer de forma automtica con la funcin **seasonal_decompose**. Enlace interesante [para series temporales](https://nwfsc-timeseries.github.io/atsa-labs/sec-tslab-decomposition-of-time-series.html)

```python
link_pax = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df_pax =  pd.read_csv(link_pax, parse_dates = ['Month'], index_col = 'Month')
df_pax.head(5)
```

Cargamos la liberas necesarias (matplotlib se utiliza para los grficos)

```python
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df_pax['Passengers'], model='additive')
result.plot()
pyplot.show()
```

Podemos acceder a las componentes por serado e intentar ver su dimensin. Ademas, podemos dibujarlo con plotly para hacerlo interactivo

```python
df_pax['observed'] = result.observed
df_pax['trend'] = result.trend
df_pax['seasonal'] = result.seasonal
df_pax['resid'] = result.resid
px.line(df_pax, template = 'plotly_white')
```

## Procesos Estacionarios


La hiptesis nula de la prueba es que la serie temporal puede ser representada por una raz unitaria, que no es estacionaria (tiene alguna estructura dependiente del tiempo). La hiptesis alternativa (que rechaza la hiptesis nula) es que la serie temporal es estacionaria.

+ **Hiptesis nula (H0)**: Si no se rechaza, sugiere que la serie temporal tiene una raz unitaria, lo que significa que no es estacionaria. Tiene una estructura dependiente del tiempo.
+ **Hiptesis alternativa (H1**): Si se rechaza la hiptesis nula, sugiere que la serie temporal no tiene una raz unitaria, lo que significa que es estacionaria. No tiene una estructura dependiente del tiempo.

Interpretamos este resultado utilizando el valor p de la prueba. Un valor p por debajo de un umbral (como el 5% o el 1%) indica que se rechaza la hiptesis nula (estacionariedad), mientras que un valor p por encima del umbral indica que no se rechaza la hiptesis nula (no estacionariedad).

+ **Valor p > 0,05**: No se rechaza la hiptesis nula (H0), los datos tienen una raz unitaria y son no estacionarios.
+ **Valor p <= 0,05**: Rechaza la hiptesis nula (H0), los datos no tienen una raz unitaria y son estacionarios.



```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df_pax['Passengers'])
print('p-value: %f' % result[1])
```

## Ruido Blanco

El ruido blanco es una funcin muy importante en series temporales. Hay muchas maneras de generar el ruido blanco, nosotros utilizaremos numpy para crear el ruido. Ojo que si no se defie una semilla, nos pueden dar funciones disintasn

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 0
std = 1 
num_samples = 1000

ruido_blanco = np.random.normal(mean, std, size=num_samples)
```

```python
pd.Series(ruido_blanco).describe()
```

```python

px.line(ruido_blanco)
```

```python
px.histogram(ruido_blanco)
```

```python
plot_acf(ruido_blanco)
plt.show()
```

```python
plot_pacf(ruido_blanco)
plt.show()
```

Curiosamente,son todos mayores que cero en valor absoluto, aunque no son estadsticamente diferentes de cero para los retrasos 1-20. Esto se debe a que estamos tratando con una muestra de las distribuciones y no con toda la poblacin de todas las variantes aleatorias. Como ejercicio, intente establecer n = 1e6 en lugar de n = 100 o n = 50 en las llamadas anteriores para generar las secuencias WN y vea qu efecto tiene en la estimacin de l ruido Tambin es importante recordarque deberamos esperar que aproximadamente 1 de cada 20 del ruido sea estadsticamente mayor que cero basndose slo en el azar, especialmente para tamaos de muestra relativamente pequeos, as que no es de extraar si alguna vez se encuentra con un caso como ste al inspeccionar los residuos del modelo.

## Random Walk

Si entendemos el concepto de random walk, podemos generarlo utilizando las mismas funciones que con el ruido blanco, utilizando la funcion de cumsum, tal como podemos apreciar

```python
# Generamos 500 pasos aleatorios
ruido_blanco = np.random.normal(loc=0, scale=1, size=500)


# Simulamos el Random Walk
random_walk = 100 + np.cumsum(ruido_blanco)

```

```python
px.line(random_walk)
```

```python
result = adfuller(random_walk)
pvalue = result[1]
```
