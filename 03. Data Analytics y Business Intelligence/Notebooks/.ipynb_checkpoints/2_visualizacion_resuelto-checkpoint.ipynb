## Importar librerias

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import plotly 
import plotly.graph_objects as go
plotly.offline.init_notebook_mode(connected = True)
```

## Leer datos

```python
df_confirmed = pd.read_csv("data/covid19_confirmed_global.csv")
df_deaths = pd.read_csv("data/covid19_deaths_global.csv")
df_recovered = pd.read_csv("data/covid19_recovered_global.csv")
```

## Tratamiento de datos

```python
id_cols = df_confirmed.columns[:4].tolist()
date_cols = df_confirmed.columns[4:].tolist()
```

```python
id_cols
```

```python
df_confirmed_long = df_confirmed.melt(
    id_vars=id_cols,
    value_vars=date_cols,
    var_name="Date",
    value_name="Confirmed"
)
df_recovered_long = df_recovered.melt(
    id_vars=id_cols,
    value_vars=date_cols,
    var_name="Date",
    value_name="Recovered"
)
df_deaths_long = df_deaths.melt(
    id_vars=id_cols,
    value_vars=date_cols,
    var_name="Date",
    value_name="Deaths"
)
```

```python
df_full = df_confirmed_long.merge(
    right=df_recovered_long,
    how="inner",
    on=["Province/State", "Country/Region", "Lat", "Long", "Date"],
    validate="1:1"
)
df_full = pd.merge(
    left=df_full,
    right=df_deaths_long,
    on=["Province/State", "Country/Region", "Lat", "Long", "Date"],
    validate="1:1"
)
```

```python
df_full.head()
```

```python
df_full.rename(columns={"Province/State": "ProvinceState", "Country/Region": "CountryRegion"}, inplace=True)
```

```python
df_full.head()
```

```python
df_full["Date"] = pd.to_datetime(df_full["Date"])
```

```python
df_full["Date"]
```

## Stacked bar plot

```python
from datetime import datetime

cols = ["CountryRegion", "Date", "Deaths"]
filter_dates = (df_full.Date >= datetime(2020, 12, 25)) & (df_full.Date < datetime(2021, 1, 1))
filter_countries = df_full.CountryRegion.isin(["Italy", "Spain", "Portugal", "Greece"])

df_stacked = df_full[(filter_dates) & (filter_countries)][cols].copy()
df_stacked.sample(5)
```

```python
filter_dates
```

```python
df_stacked.set_index(["CountryRegion", "Date"], inplace=True)
df_stacked.sort_index(inplace=True)
df_stacked.head()
```

```python
df_stacked["Deaths"][:10]
```

```python
df_stacked.groupby("CountryRegion").shift()["Deaths"][:10]
```

```python
df_stacked["DailyDeaths"] = df_stacked["Deaths"] - df_stacked.groupby("CountryRegion").shift()["Deaths"]
```

```python
df_stacked.dropna(inplace=True)
```

```python
df_stacked.reset_index(inplace=True)
```

```python
df_stacked.head()
```

### Stacked bar plot `pandas`

```python
df_pivot = df_stacked.pivot(index="Date", columns="CountryRegion", values="DailyDeaths")
df_pivot.head()
```

```python
df_pivot.plot(kind="bar", figsize=(12, 6), stacked=True, title="Daily Deaths Reported by Country", ylabel="Deaths reported")
```

### Stacked bar plot `matplotlib`

```python
fig, ax = plt.subplots(figsize=(12, 6))

bottom = np.zeros(df_stacked.Date.nunique())

for label, group in df_stacked.groupby("CountryRegion"):
    ax.bar(group.Date, group.DailyDeaths, label=label, bottom=bottom)
    bottom += group.DailyDeaths.values

ax.legend()
```

### Stacked bar plot `seaborn`

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.histplot(data=df_stacked, x="Date", hue="CountryRegion", weights="DailyDeaths", discrete=True, shrink=0.5, multiple="stack")
ax.set_title("Daily deaths by Country")
```

### Stacked bar plot `plotly.express`

```python
fig = px.bar(data_frame=df_stacked, x="Date", y="DailyDeaths", color="CountryRegion", title="Daily Deaths by Country")
fig.show()
```

### Subplots with `plotly`

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, subplot_titles=["Spain", "Italy"])

data_spain = df_multi_hist.query("CountryRegion == 'Spain'").DailyDeaths
data_italy = df_multi_hist.query("CountryRegion == 'Italy'").DailyDeaths

fig.add_trace(
    go.Histogram(x=data_spain, name="Spain", marker_color="black"),
    row=1, col=1
)
fig.add_trace(
    go.Histogram(x=data_italy, name="Italy", marker_color="black"),
    row=1, col=2,
)
fig.update_layout(title="Distribution of Daily Deaths", showlegend=False, yaxis_title="Frequency")
fig.show()
```

## Pie/Donut chart

```python
cols = ["CountryRegion", "Confirmed"]
filter_countries = df_full.CountryRegion.isin(["Italy", "Spain", "Portugal", "Greece"])

df_donut = df_full[(df_full.Date == df_full.Date.max()) & (filter_countries)][cols].copy()
df_donut.set_index("CountryRegion", inplace=True)
df_donut.head()
```

### Pie/Donut chart `pandas` and `matplotlib`
