**1) Trabajemos con listas y tuplas**
* Crea una lista y una tupla que contenga strings (al menos 3 elementos). Temática libre, sobre lo que quieras, vehículos, comida, música, etc.
* Muestra la lista y la tupla
* Muestra el 2º elemento de la lista y el penúltimo de la tupla
* Modifica (si se puede) algún elemento de la lista y de la tupla. Y mostrar el resultado
* Muestra el tamaño de la lista y de la tupla
* Realiza una búsqueda de un elemento dentro de la lista y de la tupla. Mostrar si devuelve True o False
* Añade (si se puede) algún elemento a la lista y a la tupla. Mostrar de nuevo la lista y la tupla para verificar si se ha realizado correctamente la acción.
* Borra o elimina (si se puede) el contenido de la lista y de la tupla. Mostrar de nuevo la lista y la tupla para verificar si se ha realizado correctamente la acción



```python
# Creacion
cif_restaurantes    = ("B80259690", "B56110802", "B10262376") # Tupla ya que el CIF no cambia
nombre_restaurantes  = ["VILLEMON RESTAURANTES SL", "A LA MESA GESTIONES Y EXPLOTACIONES HOSTELERAS SL", "LA COCINA DEL CHEF SL"] # Lista. El nombre puede cambiar

# Comprobar que ambas tienen el mismo número de elementos
if len(cif_restaurantes) != len(nombre_restaurantes):
    print(f'Ha ocurrido un error. El número de nombres y cif asociados son iguales')

# Mostrar
restaurantes_inscritos = len(cif_restaurantes) # Almacenar el número de restaurantes
print(f'Restaurantes inscritos a la aplicación. Total {restaurantes_inscritos}')

iterador = 0 # Iterador para mostrar la información ordenada con un único bucle
while iterador < restaurantes_inscritos:
    cif_restaurantes[iterador]
    print(f'Nombre: {nombre_restaurantes[iterador]}. CIF: {cif_restaurantes[iterador]}')
    iterador += 1
print(50 * "-")

# Mostrar elementos especificos
print(f'El segundo elemento de la lista es "{nombre_restaurantes[1]}"')
print(f'El penúltimo elemento de la tupla es "{cif_restaurantes[len(cif_restaurantes)-2]}"')
print(50 * "-")

# Modificar
nombre_restaurantes[0] = "VILLEMON RESTAURANTES 2.0 SL" # Se supone un cambio de nombre en el local
#cif_restaurantes[0] = "B80259691" # Esto producirá un TypeError. Las tuplas son inmutables
print(f'Nuveva lista de nombres de restaurantes: {nombre_restaurantes}')
print(50 * "-")

# Tamaño
print(f'El tamaño de la lista es {len(nombre_restaurantes)}')
print(f'El tamaño de la tupla es {len(cif_restaurantes)}')
print(50 * "-")

# Busqueda de elementos    
nombre_restaurante_usuario = input("Introduce el nombre del restaurante para buscar en la lista: ").upper() # Convertir todo a mayúsculas para compararlo
cif_restaurante_usuario = input("Introduce el cif del restaurante para buscar en la tupla: ").upper()       # Convertir todo a mayúsculas para compararlo

nombre_restaurante_asociado = nombre_restaurante_usuario in nombre_restaurantes # Búsqueda en lista
cif_restaurante_asociado = cif_restaurante_usuario in cif_restaurantes # Búsqueda en tupla

# Mostrar resultado de búsqueda
print(f'Búsqueda en la lista. Resultado: {nombre_restaurante_asociado}')
print(f'Búsqueda en la tupla. Resultado: {cif_restaurante_asociado}')
print(50 * "-")

# Añadir elementos
nuevo_restaurante_nombre = input("Introduce el nombre de tu restaurante para añadirlo a la lista: ").upper() # Convertir todo a mayúsculas para facilitar la búsqueda
nuevo_restaurante_cif    = input("Introduce el cif de tu restaurante para añadirla a la tupla: ").upper() 

nombre_restaurantes.append(nuevo_restaurante_nombre) # Se supone que se puede añadir un nuevo asociado a la lista
#cif_restaurantes.append(nuevo_restaurante_cif) # Esto producirá un TypeError. Las tuplas son inmutables y no tienen el método append. No es una buena solución utilizar la tupla en este caso

# Para hacerlo sobreescribimos la tupla
cif_restaurantes = tuple(list(cif_restaurantes) + [nuevo_restaurante_cif])  

print(f'Nueva lista de nombres de restaurantes: {nombre_restaurantes}')
print(f'Nueva tupla de nombres de restaurantes: {cif_restaurantes}')

print(50 * "-")

# Borrado
cif_restaurantes = () 
nombre_restaurantes = []
print(f'Nuveva lista de nombres de restaurantes: {nombre_restaurantes}')
print(f'Nuveva lista de CIF de restaurantes: {cif_restaurantes}')

print(50 * "-")
```

**2) Trabajemos con sets y diccionarios**
* Crea una set y un diccionario que contengan strings (al menos 3 elementos en el caso del set y 3 conjuntos de clave:valor en el caso del diccionario). Temática libre, sobre lo que quieras, vehículos, comida, música, etc.
* Muestra el set y el diccionario
* Muestra (si se puede) el 2º elemento del set y el valor del primer clave-valor del diccionario
* Modifica (si se puede) algún elemento del set o del diccionario. Y mostrar el resultado
* Muestra el tamaño del set y del diccionario
* Realiza una búsqueda de un elemento dentro del set y dentro del diccionario. Mostrar si devuelve True o False
* Añade (si se puede) algún elemento al set y algun clave-valor al diccionario. Mostrar de nuevo el set y el diccionario para verificar si se ha realizado correctamente la acción.
* Borra o elimina (si se puede) el contenido del set y del diccionario. Mostrar de nuevo el set y el diccionario para verificar si se ha realizado correctamente la acción



```python
NUM_GUIONES_MEDIOS = 100
print(NUM_GUIONES_MEDIOS * "-")

# Creacion
marcas_vehiculos_1 = {"Peugeot", "Ford", "Ferrari"} # Set

final_champions_league = { # Diccionario
    "year": 2020,
    "equipo": ["Paris Saint-Germain", "Bayern Munich"],
    "goles": [0, 1],
    "remates": [10, 12]
}

# Mostrar
print("Mostrar set. Elementos: ")
for marca in marcas_vehiculos_1:   # Mostrar set utilizando un bucle for. También funciona print(marcas_vehiculos_1)
    print(marca)
print(NUM_GUIONES_MEDIOS * "-")


def imprimir_diccionario(diccionario):
    """
    Función para mostrar con un formato específico el diccionario por pantalla.
    @param: diccionario: Diccionario con la información a mostrar (final_champions_league)
    """
    for key, value in diccionario.items(): # Mostrar elementos del diccionario con dos bucles (diccionario, listas).
        print(key, "\t", end="")
        if(not isinstance(value, list)): # Si no es una lista muestra el valor
            print(value)
            continue   
        for team in value:   # Cuando sea una lista itera sobre ella
            print(team, end="\t")
        print("")

print("Mostrar diccionario. Final Champions League 2020")
imprimir_diccionario(final_champions_league)
print(NUM_GUIONES_MEDIOS * "-")

    
# Mostrar elementos especificos

""" 
Los diccionarios y los set no pueden ser accedidos haciendo referencia a un índice. Por este motivo,
se utilizarán las listas para poder mostrar el elemento que dice el enunciado
"""

# Set
indice_elemento_solicitado = 1 # Índice escogido (El segundo es 2 - 1 = 1, comienza desde cero)
valor = list(marcas_vehiculos_1)[indice_elemento_solicitado] # Conversión a lista para mostrar su valor
print(f'Elemento {indice_elemento_solicitado + 1} del set escogido. Valor: {valor}') # Mostramos como el segundo elemento del set

# Diccionario
indice_clave_valor = 0
clave_solicitada = list(final_champions_league.keys())[indice_clave_valor] # Guardamos en una lista las claves del diccionario (se almacena la clave con índice introducido)
valor_solicitado = final_champions_league[clave_solicitada] # Accedemos al valor y lo almacenamos en una variable
print(f'Elemento {indice_clave_valor + 1} del diccionario escogido. Clave: {clave_solicitada}. Valor: {valor_solicitado}')
print(NUM_GUIONES_MEDIOS * "-")

# Modificar
# Set -> Una vez creado no se pueden modificar los elementos del set. Pero se podría crear un nuevo set en el mismo espacio de memoria con los nuevos elementos (como primer ejercicio)

# Diccionario
# Final temporada 2020-2021
final_champions_league['year'] = 2021 
final_champions_league['equipo'] = ['nc', 'nc']
final_champions_league['goles'] = ['nc', 'nc']
final_champions_league['remates'] = ['nc', 'nc'] 

# Mostrar
print('Diccionario modificado. Final (2020-2021):')
imprimir_diccionario(final_champions_league)
print('\nNota: "nc = no conocido"')
print(NUM_GUIONES_MEDIOS * "-")

# Tamaño
print(f'El tamaño del set es {len(marcas_vehiculos_1)}')
print(f'El tamaño del diccionario es {len(final_champions_league)}')
print(NUM_GUIONES_MEDIOS * "-")

# Busqueda de elementos
# Set
busqueda_set = input("Introudce una marca de vehículos para comprobar si se encuentra almacenada dentro del set: ")
resultado_busqueda_set = busqueda_set in marcas_vehiculos_1
print(f'El campo introducido ({busqueda_set}) es válido: {resultado_busqueda_set}')

# Diccionario
busqueda_diccionario = input("Introduce un campo a buscar en el diccionario: ")
resultado_busqueda_diccionario = busqueda_diccionario in final_champions_league
print(f'El campo introducido ({busqueda_diccionario}) es válido: {resultado_busqueda_diccionario}')
print(NUM_GUIONES_MEDIOS * "-")

# Añadir elementos
# Set: Dos métodos:add() y update()
marcas_vehiculos_1.add("BMW")
print(f'Nuevo set extendido (add()): {marcas_vehiculos_1}')

marcas_vehiculos_1.update(["Mercedes", "Audi", "Peugeot"])
print(f'Nuevo set extendido (update()): {marcas_vehiculos_1}')  # Como es un set, no existen elementos repetidos y Peugeot solo aparecerá una vez

# Diccionario
"""
En las estadísticas de 2021 se contabilizarán los campos anteriores y además:
- las tarjetas amarillas
- las faltas recibidas
"""
final_champions_league["Tarjetas amarillas"] = ["nc", "nc"] 
final_champions_league["Faltas recibidas"] = ["nc", "nc"]   
print('\nDiccionario modificado. Final (2020-2021):')
imprimir_diccionario(final_champions_league)
print('\nNota: "nc = no conocido"')
print(NUM_GUIONES_MEDIOS * "-")

# Borrado
marcas_vehiculos_1.clear() # Eliminar todo el contenido del set
final_champions_league.clear() # Eliminar todo el contenido del diccionario
print(f'Se ha borrado el set. Contenido: {marcas_vehiculos_1}')
print(f'Se ha borrado el diccionario. Contenido: {final_champions_league}')

print(NUM_GUIONES_MEDIOS * "-")
```

**3) Realizar un programa que pida al usuario 3 números que pueden ser flotantes (no es necesario realizar bucles aún, podemos repetir el código), estos números se deberán introducir en una lista. Cuando se haya finalizado la introducción de los datos, se mostrará el sumatorio de toda la lista. Guardar el sumatorio en una variable que se llame "sumatorio"**

*Ayuda: Existe una función llamada sum(lista) ¡Pruébalo!*

```python
numeros_usuario = []
NUM_ELEMENTOS = 3
sumatorio = 0
while len(numeros_usuario) < NUM_ELEMENTOS:
    valor_introducido = float(input("Introduce un número real: "))
    numeros_usuario.append(valor_introducido)
    sumatorio += valor_introducido
print(f'El sumatorio de los datos introducidos es: {sumatorio}')
print(f'El sumatorio de los datos introducidos (función "sum()") es: {sum(numeros_usuario)}')
```

**5) Dada la siguiente matriz (lista con cuatro listas anidadas) se observa que en cada fila, el cuarto elemento siempre debe ser el resultado de sumar los tres primeros. 
Pista: Modifica las sumas erróneas utilizando slicing**

* No vale poner directamente los valores (ni los índices ni la suma):
* matriz[1][3] = 6
* matriz[3][3] = 12
* Hay que buscar una forma automatizada de acceder y modificar esos valores

```python
matriz = [ 
    [1, 1, 1, 3],
    [2, 2, 2, 7],
    [3, 3, 3, 9],
    [4, 4, 4, 13]
]

"""
La matriz correcta es

    [1, 1, 1, 3],
    [2, 2, 2, 6], # Cambiar elemento [1][3] = 7 --> 6
    [3, 3, 3, 9],
    [4, 4, 4, 12] # Cambiar elemento [3][3] = 13 --> 12
"""

# Iteradores
r = 0 # row
c = 0 # column

# Recorrer lista
for fila in matriz:
    valores = []
    # Añadir todos los elementos de la fila a una lista
    for columna in fila:
        valores.append(columna)
        c+=1 # Incrementar valor columna
        
    sumatorio = sum(valores[:len(fila) - 1]) # Obtener sumatorio de los elementos (excepto el último)
    
    # Comprobar que el cuarto elemento (último) de cada fila es el el resultado de sumar los tres primeros.
    if(matriz[r][-1] != sumatorio):
        matriz[r][-1] = sumatorio
    r+=1 # Incrementar valor fila
    
print(f'La matriz resultado es: {matriz}')
        
```

**6) Supongamos que obtenemos una cadena de texto corrupta, esta al revés. Esta cadena contiene el nombre y la nota de un alumno. Formatear dicha cadena para obtener el siguiente formato:**

* ***Nombre*** ***Apellido*** ha sacado un ***Nota*** de nota.

*Pista: Para voltear una cadena usando técnicas de slicing se puede utilizar un tercer índice -1: cadena[::-1]*

```python
cadena = "zereP solraC,01"

def reparar_cadena(cadena):
    """
    Repara la cadena corrupta (está al revés) que contiene el nombre y la nota de un alumno para obtener
    el siguiente formato
    
    Formato: Nombre Apellido ha sacado un Nota de nota.
    
    @param: cadena: Cadena corrupta
    @return: formato_alumno: Cadena reparada
    """
    # Voltear cadena
    cadena_modificada = cadena[::-1]

    # Utilizar split para identificar las dos partes de la cadena (NOTA, NOMBRE APELLIDO)
    datos_alumno = cadena_modificada.split(",")
    
    # Obtener el formato correcto
    formato_alumno = datos_alumno[1] + " ha sacado un " + datos_alumno[0] +" de nota."
    
    # Devolver formato
    return formato_alumno

cadena_reparada = reparar_cadena(cadena)
print(cadena_reparada)
```

**7) Realizar un programa que cumpla las siguientes instrucciones:**
* Crear un conjunto llamado usuarios con los usuarios Marta, David, Elvira, Juan y Marcos
* Crear un conjunto llamado administradores con los administradores Juan y Marta.
* Borra al administrador Juan del conjunto de administradores.
* Añade a Marcos como un nuevo administrador, pero no lo borres del conjunto de usuarios.
* Muestra todos los usuarios por pantalla de forma dinámica, además debes indicar cada usuario es administrador o no.

*Notas: Los conjuntos se pueden recorrer dinámicamente utilizando el bucle for de forma similar a una lista. También cuentan con un método llamado **.discard(elemento)** que sirve para borrar un elemento.* 

```python
# Crear conjuntos
usuarios = {'Marta', 'David', 'Juan', 'Marcos'} # usuarios
administradores = {'Juan', 'Marta'} # Administradores

# Borrar el administrador Juan del conjunto
administradores.remove('Juan')

# Añadir a marcos como nuevo administrador sin borrarlo del conjunto de usuarios.
administradores.add('Marcos')

for user in usuarios:
    print(user, end=" ")
    if(user in administradores):
        print("es administrador")
    else:
        print("no es administrador")

```

**8) Realizar un programa basado en un videojuego que tiene determinados jugadores con características a nivel 2 de forma predeterminada. Se deben cumplir las siguientes condiciones:**
* El caballero tiene el doble de vida y defensa que un guerrero.
* El guerrero tiene el doble de ataque y alcance que un caballero.
* El arquero tiene la misma vida y ataque que un guerrero, pero la mitad de su defensa y el doble de su alcance.
* Muestra como quedan las propiedades de los tres personajes.

```python
# Caracteristicas de los personajes de nivel 1
caballero = { 'vida':2, 'ataque':2, 'defensa': 2, 'alcance':2 }
guerrero  = { 'vida':2, 'ataque':2, 'defensa': 2, 'alcance':2 }
arquero   = { 'vida':2, 'ataque':2, 'defensa': 2, 'alcance':2 }


"""
Para subir al nivel dos se crea una copia del nivel 1 que se almacenará en variables (_nivel1) para mantener un registro de
los niveles que han tenido los personajes
"""

# Copias de seguridad.
caballero_nivel1 = caballero.copy()
guerrero_nivel1  = guerrero.copy()
arquero_nivel1   = arquero.copy()

# Modificaciones al subir al nivel 2.
# (1) El caballero tiene el doble de vida y defensa que un guerrero
caballero['vida']    = 2 * guerrero['vida']
caballero['defensa'] = 2 * guerrero['defensa']

# (2) El guerrero tiene el doble de ataque y alcance que un caballero
guerrero['ataque']  = 2 * caballero['ataque']
guerrero['alcance'] = 2 * caballero['alcance']

# (3) El arquero tiene la misma vida y ataque que un guerrero, pero la mitad de su defensa y el doble de su alcance
arquero['vida']    = guerrero['vida']
arquero['ataque']  = guerrero['ataque']
arquero['defensa'] = int(0.5 * guerrero['defensa']) # Suponemos que solo valen números enteros
arquero['alcance'] = 2   * guerrero['alcance']

def mostrar_propiedades(personajes):
    """
    Imprime por pantalla las propiedades de los jugadores disponibles.
    """
    for key,value in personajes.items():
        print(key, end="\t")
        for key, value in value.items():
            print(key + ": " + str(value), end="\t")
        print("\n")
    
personajes = {'caballero': caballero, 
              'guerrero':  guerrero, 
              'arquero':   arquero
             }

print(50 *  '-')
print("Personajes Nivel 2: ")
print(50 *  '-')
mostrar_propiedades(personajes)
print(50 *  '-', end="\n\n")

print(50 *  '-')
print("Personajes Nivel 1: ")
print(50 *  '-')

# Cargamos la copia de seguridad para mostrar el nivel anterior
personajes['caballero'] = caballero_nivel1 
personajes['guerrero']  = guerrero_nivel1 
personajes['arquero']   = arquero_nivel1 

# Mostrar propiedades
mostrar_propiedades(personajes)
print(50 *  '-')
```
