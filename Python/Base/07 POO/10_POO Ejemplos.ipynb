# Clases y objetos (ejemplos)


## Ejemplo Vehiculos

```python
class Vehiculo():
    ''' 
        Clase Vehículos. Incluye la marca y model de un vehículo.
        Por defecto, el vehículo no está en marcha, ni acelerando ni frenando
    '''
    
    # Constructor
    def __init__(self, marca, modelo):
        """ Constructor del Vehículo (marca y modelo)"""
        self.marca = marca
        self.modelo = modelo
        self.enmarcha = False
        self.acelera = False
        self.frena = False
        
    # Métodos de la clase
    def arrancar(self):
        '''Método arrancar() de la clase Vehículos. Indica al sistema que el vehículo está en marcha'''
        self.enmarcha = True
        
    def acelerar(self):
        ''' Método acelerar() de la clase Vehículos. Indica al sistema que el vehículo está acelerando'''
        self.acelera = True

    def frenar(self):
        ''' Método frenar() de la clase Vehículos. Indica al sistema que el vehículo está frenando.'''
        self.frena = True
    
    def __str__(self):
        ''' Indica el estado del vehículo '''
        return 'Marca: {}. Modelo: {}. En marcha: {}. Acelera: {}. Frena: {}'.format(self.marca,self.modelo,self.enmarcha,self.acelera,self.frena)
    
class Moto(Vehiculo):
    ''' Clase Moto que hereda de la clase Vehiculos '''
    # Atributos de la clase
    hacer_caballito = ''
    
    def caballito(self):
        ''' Método caballito() de la clase Moto. Indica al sistema si la moto está haciendo un caballito'''
        self.hacer_caballito = 'Voy haciendo un caballito'
        
    # Sobreescribimos el método __str__ (overriding)
    def __str__(self):
        ''' Método de la clase heredado de Vehículos. '''
        return 'Marca: {}. Modelo: {}. En marcha: {}. Acelera: {}. Frena: {}. Caballito: {}'.format(self.marca,self.modelo,self.enmarcha,self.acelera,self.frena, self.hacer_caballito)

    
class Furgoneta(Vehiculo):
    """ Clase Furgoneta que hereda de Vehiculo """

    # Atributos
    cargado = False
    
    def carga(self, cargar):
        """ Método carga() de la clase Furgoneta. Indica al sistema si la furgoneta está cargada o no """
        self.cargado = cargar
        
        if self.cargado:
            return 'La furgoneta está cargada'
        else:
            return 'La furgoneta no está cargada'
        
class VElectrico():
    """ Clase VElectrico"""
    # Atributos
    autonomia = 0
    cargando  = False
    
    # Constructor
    def __init__(self):
        self.autonomia = 100

    def cargarEnergia(self):
        '''Método cargarEnergia() de la clase VElectrico. Indica al sistema si el vehículo está cargando o no.'''
        self.cargando = True
        
class BicicletaElectrica(VElectrico, Vehiculos):
    """ Clase BicicletaElectrica que hereda de la clase Vehículos y de la clase VElectricos, es decir,
        HERENCIA MÚLTIPLE
    """
    pass
```

```python
# Moto
print('\n\n--- Moto ---')
miMoto = Moto('Derbi', 'DRD Racing')
print(miMoto)
miMoto.caballito()
print(miMoto)
print('---')

# Furgoneta
print('\n\n--- FURGONETA ---')
miFurgoneta = Furgoneta('Renault', 'Kangoo')
miFurgoneta.arrancar()
print(miFurgoneta)
estadoFurgonetaCargada = miFurgoneta.carga(True)
print('Cargado:',estadoFurgonetaCargada)
print('---')

# Bicicleta Eléctrica
print('\n\n--- BICICLETA ELÉCTRICA ---')
# Al heredar de dos clases, tiene disponible dos constructores. ¿Cuál ejecuta?
# Por defecto, ejecuta el constructor de la primera clase.
miBici = BicicletaElectrica()
print('Autonomía:',miBici.autonomia)
print('---')
```

## Ejemplo Persona-Empleado

```python
class Persona():
    """ Clase Persona """
    # Atributos
    nombre = ''
    edad = 0
    lugar_residencia = ''
    
    # Constructor
    def __init__(self, nombre, edad, lugar_residencia):
        """ Constructor bicicleta """
        self.nombre = nombre
        self.edad = edad
        self.lugar_residencia = lugar_residencia
        
    # Método de descripción
    def descripcion(self):
        """ Método que se ejecuta al utilizar print() """
        print('Nombre: {}\nEdad: {}\nLugar de residencia: {}'.format(self.nombre, self.edad, self.lugar_residencia))
        
class Empleado(Persona):
    """ Clase Empleado que hereda de la clase Persona"""
    # Atributos
    salario = 0
    antiguedad = 0
    
    # Constructor extendido
    def __init__(self, salario, antiguedad, nombre_empleado, edad_empleado, residencia_empleado):
        """ Constructor extendendido de Empleado que hereda de Persona """
        # Con el método super() se hace referencia a la clase padre (Persona)
        # super().__init__ hace referencia al constructor de la clase padre.
        # Llamada al constructor de la clase padre
        super().__init__(nombre_empleado, edad_empleado, residencia_empleado)
        self.salario = salario
        self.antiguedad = antiguedad

    def descripcion(self):
        """ Método que se ejecuta al realizar el print()"""
        super().descripcion()
        print('Salario: {}\nAntiguedad (años): {}'.format(self.salario, self.antiguedad))    
```

```python
# Crear empleado
Manuel = Empleado(20000, 3, 'Manuel', 55, 'Madrid')
Manuel.descripcion()
```
