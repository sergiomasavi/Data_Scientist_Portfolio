**1) Realiza una función que devuelva el área de un rectángulo a partir de su base y su altura. La función se llamará area_rectangulo(). La base y la altura se pedirán al usuario por teclado.**

*Nota: El área de un rectángulo se obtiene al multiplicar la base por la altura.*

````python
import sys
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np

SEPARADOR = 50 * '-'

def area_rectangulo(rectangulo: dict) -> dict:
    """
    Función que devuelve el área de un rectángulo a partir de su base y su altura.
    
    :param altura: Altura del rectángulo
    :param base: Base del rectángulo
    
    :return area: area = base * altura
    """
    rectangulo['area'] = rectangulo['base'] * rectangulo['altura']
    return rectangulo

def definir_rectangulo():
    """
    Función que se encarga de interaccionar con el usuario
    
    :return dimensiones = {base, altura}. Dimensiones del rectángulo
    """
    
    # Valor por defecto de la base y la altura del triángulo 
    base = None
    altura = None
    
    # Mensaje a mostrar en la recepción de datos por teclado
    message = 'Introduce la base y altura del rectángulo separado por comas. Ej: 5,5'
    
    # Bucle infinito para pedir dos números por teclado
    while True:
        # Pedir datos por teclado
        valor_str = input(message + ': ')
    
        # Split la cadena con la coma (caracter de identificación)
        valor_float = valor_str.split(',')

        # Comprobar que tiene dos elementos (base y altura)
        if len(valor_float) == 2:
            try:
                # Guardar valores
                base = float(valor_float[0])
                altura = float(valor_float[1])
                
                # Romper bucle infinito
                break
                
            except ValueError as e:
                print('ValueError:',e,'\nIntroduce valores numéricos', file=sys.stderr)
                print(SEPARADOR)
                continue
        
    # Variable de salida
    dimensiones = {'base':   base,
                   'altura': altura
                  }

    # Devolución
    return dimensiones


def plotRectangulo(rectangulo:dict) -> None:
    """
    Representa un triángulo
    """
    fig = plt.figure() 
    ax = fig.add_subplot(111) 
    rect_plot = matplotlib.patches.Rectangle((0,0), rectangulo['base'], rectangulo['altura'], angle=0.0, color ='green')
    ax.add_patch(rect_plot)

    # Definir titulo
    plt.title('Rectángulo')

    # Definir grid
    ax.grid(color='gray', ls = '-.', lw = 0.25)

    # Establecer límites
    xlim_abs = rectangulo['base'] + 1
    ylim_abs = rectangulo['altura'] + 1

    # Definir ejes
    plt.xlim(-xlim_abs, xlim_abs)
    plt.ylim(-ylim_abs, ylim_abs)

    plt.xlabel('Base (mm)', fontsize=18)
    plt.ylabel('Altura (mm)', fontsize=18)

    plt.show() 
    
    
    
# Función main
if __name__ == '__main__':
    # Pedir al usuario que defina el rectángulo
    rectangulo = definir_rectangulo()
    print(SEPARADOR)

    # Calcular área del rectángulo
    rectangulo = area_rectangulo(rectangulo)

    # Mostrar resultado
    print('Un rectángulo con base {:.2f} mm y altura {:.2f} mm tiene un área de {:.2f} mm'.format(rectangulo['base'], 
                                                                                         rectangulo['altura'],
                                                                                         rectangulo['area']))
    print(SEPARADOR)

    # Extra. Dibujar rectángulo.
    print('Dibujando rectángulo en (0,0)')
    plotRectangulo(rectangulo)
````

**2) Realiza una función que devuelva el área de un círculo a partir de un radio. La función se llamará area_circulo(). El radio se pedirá al usuario por teclado:**

Nota: El área de un círculo se obtiene al elevar el radio a dos y multiplicando el resultado por el número pi. Puedes utilizar el valor 3.14159 como pi o importarlo del módulo math:
```python

print(math.pi)
> 3.1415...
```

````python
import sys
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math

SEPARADOR = 50 * '-'

def area_circulo(circulo: dict) -> dict:
    """
    Función que devuelve el área de un rectángulo a partir de su base y su altura.
    
    :param altura: Altura del rectángulo
    :param base: Base del rectángulo
    
    :return area: area = base * altura
    """
    circulo['area'] = pow(circulo['radio'], 2) * math.pi
    
    return circulo

def definir_circulo():
    """
    Función que se encarga de interaccionar con el usuario
    
    :return circulo = {radio}. Radio del círculo
    """
    # Mensaje a mostrar en la recepción de datos por teclado
    message = 'Introduce el radio del círculo: '
    
    # Bucle infinito para pedir dos números por teclado
    while True:
        # Pedir datos por teclado
        valor_str = input(message)
        try:
            # Guardar valor
            radio = float(valor_str)

            # Variable de salida
            circulo = {'radio':   radio}

            # Devolución
            return circulo

        except ValueError as e:
            print('ValueError:',e,'\nIntroduce un valor numérico', file=sys.stderr)
            print(SEPARADOR)
            continue
        


def plotCirculo(circulo:dict) -> None:
    """
    Representa un triángulo
    """
    fig = plt.figure() 
    ax = fig.add_subplot(111) 
    circle_plot = matplotlib.patches.Circle((0,0), circulo['radio'], color ='green')
    ax.add_patch(circle_plot)

    # Definir titulo
    plt.title('Círculo')

    # Definir grid
    ax.grid(color='gray', ls = '-.', lw = 0.25)

    # Establecer límites
    xlim_abs = circulo['radio'] + circulo['radio']
    ylim_abs = xlim_abs

    # Definir ejes
    plt.xticks(np.arange(-xlim_abs, xlim_abs, step=1))
    plt.yticks(np.arange(-xlim_abs, xlim_abs, step=1))

    plt.xlim(-xlim_abs, xlim_abs)
    plt.ylim(-ylim_abs, ylim_abs)

    plt.xlabel('Ancho (mm)', fontsize=18)
    plt.ylabel('Alto (mm)', fontsize=18)

    plt.show() 

# Función main
if __name__ == '__main__':
    # Pedir al usuario que defina el circulo
    circulo = definir_circulo()
    print(SEPARADOR)

    # Calcular área del circulo
    circulo = area_circulo(circulo)

    # Mostrar resultado
    print('Un círculo con radio {:.2f} mm tiene un área de {:.2f} mm'.format(circulo['radio'], 
                                                                       circulo['area']))
    print(SEPARADOR)

    # Extra. Dibujar rectángulo.
    print('Dibujando círculo en (0,0)')
    plotCirculo(circulo)
````

**3) Realiza una función que a partir de dos números que se solicitan por teclado al usuario y que se llamará relacion(), cumpla lo siguiente:**

* Si el primer número es mayor que el segundo, debe devolver 1.
* Si el primer número es menor que el segundo, debe devolver -1.
* Si ambos números son iguales, debe devolver un 0.

**Comprueba la relación entre los números: '5 y 10', '10 y 5' y '5 y 5'**

````python
import sys

SEPARADOR = 50 *'-'

def relacion(numeros):
    """
    Analiza la relación entre dos números.
    
    :param numeros: Lista con los dos números a evaluar numeros = [num0, num1]
    """
    num0 = numeros[0]
    num1 = numeros[1]
    if num0 > num1:
        return 1
    elif num0 < num1:
        return -1
    else:
        return 0
    
def pedir_numeros(message):
    """
    Función que se encarga de interaccionar con el usuario
    
    :return numeros = [num0, num1]: Números introducidos por el usuario
    """
    
    
    # Bucle infinito para pedir dos números por teclado
    while True:
        # Pedir datos por teclado
        valor_str = input(message)
    
        # Split la cadena con la coma (caracter de identificación)
        valor_str = valor_str.split(',')

        # Comprobar que tiene dos elementos (base y altura)
        if len(valor_str) == 2:
            try:
                # Devolver número
                numeros = [float(valor_str[0]), float(valor_str[1])]
                return numeros
                
            except ValueError as e:
                print('ValueError:',e,'\nIntroduce valores numéricos', file=sys.stderr)
                print(SEPARADOR)
                continue

                
if __name__ == '__main__':
    # Para probar el ejercicio, solicitar tres pares de números (5,10), (10,5) y (5,5)
    iter = 0 # Contador con el número iteraciones de la aplicación
    NUM_MAX = 3 # Limitador de iteraciones
    
    message = 'Introduce dos números separados por ",": '
    
    # Bucle iterativo
    while iter < NUM_MAX:
        numeros = list() # Inicializar variable para almacenar números (cc. TypeError en segunda iteración)
        # Solicitar números
        numeros = pedir_numeros(message)
        
        # Evaluar
        rel = relacion(numeros)
        
        # Mostrar mensaje por pantalla
        print('Para los números {} y {} se obtiene la relacion {}'.format(numeros[0], numeros[1], rel))
        print(SEPARADOR)
        
        iter+=1
````

**4) Realiza una función llamada lee_numero() que solicite y lea por teclado un numero. Utilizar esta función para pedirle al usuario 3 números. Luego pasarle estos 3 números a una función que se llame mayor() que tenga 3 parámetros y que devuelva el número mayor de los 3:**

````python
import sys

SEPARADOR = 50 * '-'

def pedirNumero()->int:
    """
    Pedir número al usuario (casting a float)
    :return: num (float)
    """

    while True:  # Bucle para pedir número
        try:
            # No se han obtenido aún los tres números del usuario. Pedirle la información y almacenarla.
            num = float(input('Introduce un valor númerico: '))
            return num
        except TypeError:
            print('El número introducido no es correcto', file=sys.stderr)
            continue

def lee_numero(numeros: list, N: int) -> list:
    """
    Función recursiva (para practicar) que recibe una lista y evalúa su longitud. Mientras que no se alcance la longitud de
    tres números, no dejará de llamarse a ella misma pidiendo el número al usuario y almacenandolo en una lista.
    
    :param numeros: Lista de números obtenidos
    :return numeros: Lista de números obtenidos.
    """

    if len(numeros) != N:
        # Obtener número
        num = pedirNumero()

        # Insertar número en la lista de números
        numeros.append(num)

        # LLamar de nuevo a la función recursiva
        numeros = lee_numero(numeros, N)

    if len(numeros) == 3:
        return numeros
    
def mayor(num0: int, num1: int, num2: int) -> int:
    """
    Obtiene el valor máximo de una lista de números (longitud = 3). max().
    
    
    Nota. Se pasan tres números para seguir el enunciado del ejercicio. Sin embargo, se propone que la función tenga un único
    parámetro (lista de números) para reducir el número de variables.
    
    :param num0: Primer número obtenido.
    :param num1: Segundo número obtenido.
    :param num2: Tercer número obtenido.
    
    :return mayor: Valor máximo de la lista
    """
    
    # Guardar en una lista
    numeros = [num0, num1, num2]
    
    
    num_mayor = numeros[0] # El mayor número por defecto es el primer elemento
    
    for indice in range(1, len(numeros)): # Bucle para recorrer la lista por índices
        if numeros[indice] > num_mayor:
            num_mayor = numeros[indice]
            
    return num_mayor
        
if __name__ == '__main__':
    
        # Obtener lista de 3 números,
        N = 3 # Cantidad de números
        numeros = lee_numero(list(), N)
        

        # Obtener número mayor de la lista.
        #mayor = max(numeros) # Opción directa
        num_mayor = mayor(numeros[0], numeros[1], numeros[2]) 
        print(SEPARADOR)
        print('El número mayor es', num_mayor)
````

**5) Realizar una función llamada imc(). Esta función nos proporcionará el estado nutricional de una persona. Esto es; la clasificación de su índice de masa corporal (IMC), según la tabla proporcionada. La función recibirá el peso (en kilos) y la talla de una persona (en metros con dos decimales), datos a partir de los cuales se puede calcular el IMC cómo:**

IMC = peso / (altura * altura)

Clasificación: IMC (Kg/m^2)
* Bajo peso:  < 18.50
* Normal:      [18.50, 25.00)
* Sobrepeso: >= 25.00
* Obesidad: >= 30.00

````python
import sys

SEPARADOR = 50 * '-'

def imc(peso: float, altura: float) -> str:
    """
    Obtener el estado nutricional de una persona (IMC).
    
    :param peso: Peso de la persona en kg (float)
    :param talla: Altura de la persona en m (float con dos decimales)
    """
    # Calcular Indice de Masa Corporal
    IMC = peso / pow(altura, 2)
    
    if IMC < 0:
        estado = 'Error en la entrada de información'
    elif IMC < 18.50:
        estado = 'Bajo peso'
    elif 18.50 <= IMC < 25.00:
        estado = 'Normal'
    elif 25 <= IMC < 30:
        estado = 'Sobrepeso'
    else:
        estado = 'Obesidad'
        
    return estado
    
if __name__ == '__main__':
    
    # En esta ocasión no se utilizará la entrada por teclado.
    peso  = 80.5 # Peso de la persona (kg)
    talla = 1.90   # Altura de la persona (m)
    
    # Obtener estado
    estado = imc(peso, talla) 
    
    print('Resumen del estado del cliente:\nAltura: {:.2f}\nPeso: {:.2f}\nEstado Nutricional: {}'.format(talla, peso, estado))
````
