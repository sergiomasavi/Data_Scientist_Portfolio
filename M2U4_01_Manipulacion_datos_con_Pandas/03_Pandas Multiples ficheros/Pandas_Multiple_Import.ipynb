# Ejemplo de unir múltiples CSV

```python
import pandas as pd
import glob

path = r'./' 
all_files = glob.glob(path + "/*.csv")

lista = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    lista.append(df)

frame = pd.concat(lista, axis=0, ignore_index=True)
frame
```

# Ejemplo de unir múltiples excels

```python
import os
import pandas as pd

# path = os.getcwd() # Detectar automaticamente la ruta
path = "./"
files = os.listdir(path)

# Filtramos por los ficheos xlsx
files_xls = [f for f in files if f[-4:] == 'xlsx']
print("Ficheros excel encontrados: ", files_xls)

# Inicializamos un dataframe vacio
df = pd.DataFrame()

for f in files_xls:
    data = pd.read_excel(f) # pd.read_excel(f, 'Hoja1')
    df = df.append(data).reset_index(drop = True)
    
df
```
