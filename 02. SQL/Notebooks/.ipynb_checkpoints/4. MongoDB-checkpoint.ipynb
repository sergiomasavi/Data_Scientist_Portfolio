## AGREGACIONES

Necesitan tres parámetros: 
- 1. (dolar)match: es el equivalmente al where de SQL.
- 2. (dolar)group: los campos sobre los que vamos a agrupar (_id será la clave del campo que vamos a agrupar).
- 3. (dolar)sort: los campos por los que vamos a ordenar
Se los pasamos en una lista: 


```python
# Primero, definimos los parámetros:
lParams = [
    {
        # GROUP BY
        "$group": {
            "_id": "$nombre",   # _id código interno de MongoDB para indicar en el valor, el nombre del campo sobre el que vamos a agrupar.
            "Count": {"$sum": 1},
            "sum_rating":{"$sum":"$rating"},
            "Mean_rating":{"$avg":"$rating"},
            "Max_rating":{"$max":"$rating"},
            "Min_rating":{"$min":"$rating"},
        }
    }, 
    {
        # ORDER BY
        "$sort": {
            "_id": 1 # estamos ordenando de forma ascendente (1) el campo $_id que es el "rating"
        }
    }
]
```

```python
db.reviews.aggregate(lParams)
```

```python
pd.DataFrame(data=list(db.reviews.aggregate(lParams)))
```

```python
results=pd.DataFrame(data=list(db.reviews.aggregate(lParams)))
results
```

```python
results.set_index('_id')['Mean_rating'].plot()
```

```python
# Primero, definimos los parámetros:
lParams = [
    {
        # Where
        "$match": {
            "$or": [ { "nombre": "KFC" }, {"nombre": "Burger King"} ]
        }
    }, 
    {
        # GROUP BY
        "$group": {
            "_id": "$nombre",   # _id código interno de MongoDB para indicar en el valor, el nombre del campo sobre el que vamos a agrupar.
            "Count": {"$sum": 1},
            "sum_rating":{"$sum":"$rating"},
            "Mean_rating":{"$avg":"$rating"},
            "Max_rating":{"$max":"$rating"},
            "Min_rating":{"$min":"$rating"},
        }
    }, 
    {
        # ORDER BY
        "$sort": {
            "_id": 1 # estamos ordenando de forma ascendente (1) el campo $_id que es el "rating"
        }
    }
]
```

```python
pd.DataFrame(data=list(db.reviews.aggregate(lParams)))
```
